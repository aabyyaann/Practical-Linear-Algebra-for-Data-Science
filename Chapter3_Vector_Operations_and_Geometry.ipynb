## 2. Menyiapkan Beberapa Vektor Contoh

Pada chapter ini kita akan menggunakan beberapa vektor kecil agar konsepnya mudah dipahami secara visual dan numerik.

```python
v = np.array([2, 1])
w = np.array([1, 3])

print("Vektor v =", v)
print("Vektor w =", w)
```

## 3. Penjumlahan Vektor

Jika dua vektor memiliki jumlah elemen yang sama, kita dapat menjumlahkannya secara elemen per elemen.

Contoh:
v = [2, 1]
w = [1, 3]

maka:
v + w = [2+1, 1+3] = [3, 4]

Secara geometris:
- vektor hasil penjumlahan menunjukkan gabungan perpindahan dari dua vektor.

```python
v_plus_w = v + w

print("v =", v)
print("w =", w)
print("v + w =", v_plus_w)
```

## 4. Visualisasi Penjumlahan Vektor

Kita bisa memvisualisasikan vektor `v`, `w`, dan hasil `v + w` pada bidang 2D.

Interpretasi:
- `v` adalah perpindahan pertama
- `w` adalah perpindahan kedua
- `v + w` adalah hasil gabungan keduanya

```python
plt.figure(figsize=(7, 7))
plt.axhline(0, color="gray", linewidth=1)
plt.axvline(0, color="gray", linewidth=1)

plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, label='v')
plt.quiver(0, 0, w[0], w[1], angles='xy', scale_units='xy', scale=1, label='w')
plt.quiver(0, 0, v_plus_w[0], v_plus_w[1], angles='xy', scale_units='xy', scale=1, label='v + w')

plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.xlabel("Sumbu X")
plt.ylabel("Sumbu Y")
plt.title("Visualisasi Penjumlahan Vektor")
plt.legend()
plt.grid(True)
plt.show()
```

## 5. Pengurangan Vektor

Pengurangan vektor juga dilakukan elemen per elemen.

Contoh:
v = [2, 1]
w = [1, 3]

maka:
v - w = [2-1, 1-3] = [1, -2]

Secara geometris, `v - w` dapat dipahami sebagai arah dari `w` menuju `v`.

```python
v_minus_w = v - w

print("v =", v)
print("w =", w)
print("v - w =", v_minus_w)
```

## 7. Perkalian Skalar

Perkalian skalar berarti mengalikan seluruh elemen vektor dengan satu angka.

Misalnya:
v = [2, 1]

Jika dikalikan 3:
3v = [6, 3]

Jika dikalikan 0.5:
0.5v = [1, 0.5]

Jika dikalikan -1:
-v = [-2, -1]

Makna geometris:
- skalar > 1 memperpanjang vektor
- 0 < skalar < 1 memperpendek vektor
- skalar negatif membalik arah vektor

```python
v2 = np.array([2, 1])

scaled_2 = 2 * v2
scaled_half = 0.5 * v2
scaled_neg = -1 * v2

print("v =", v2)
print("2v =", scaled_2)
print("0.5v =", scaled_half)
print("-v =", scaled_neg)
```

## 8. Visualisasi Perkalian Skalar

Kita bandingkan beberapa hasil skalar terhadap vektor yang sama.

```python
plt.figure(figsize=(8, 8))
plt.axhline(0, color="gray", linewidth=1)
plt.axvline(0, color="gray", linewidth=1)

vectors = {
    "v": v2,
    "2v": scaled_2,
    "0.5v": scaled_half,
    "-v": scaled_neg
}

for name, vec in vectors.items():
    plt.quiver(0, 0, vec[0], vec[1], angles='xy', scale_units='xy', scale=1, label=name)

plt.xlim(-4, 5)
plt.ylim(-4, 5)
plt.xlabel("Sumbu X")
plt.ylabel("Sumbu Y")
plt.title("Visualisasi Perkalian Skalar pada Vektor")
plt.legend()
plt.grid(True)
plt.show()
```

## 9. Panjang Vektor (Norm)

Panjang atau besar vektor sering disebut **norm**.

Untuk vektor dua dimensi:
v = [x, y]

panjangnya:
||v|| = sqrt(x^2 + y^2)

Untuk dimensi lebih tinggi, prinsipnya sama:
akar dari jumlah kuadrat semua elemen.

```python
a = np.array([3, 4])

norm_manual = np.sqrt(a[0]**2 + a[1]**2)
norm_numpy = np.linalg.norm(a)

print("Vektor a =", a)
print("Norm manual =", norm_manual)
print("Norm NumPy =", norm_numpy)
```

## 10. Menghitung Norm Beberapa Vektor

Agar lebih rapi, kita buat tabel beberapa vektor dan panjangnya.

```python
vectors = {
    "v1": np.array([3, 4]),
    "v2": np.array([1, 1]),
    "v3": np.array([5, 12]),
    "v4": np.array([-2, 6]),
    "v5": np.array([0, 7])
}

rows = []
for name, vec in vectors.items():
    rows.append({
        "Nama": name,
        "Vektor": str(vec),
        "Norm": np.linalg.norm(vec)
    })

norm_df = pd.DataFrame(rows)
norm_df
```

## 12. Dot Product

Dot product adalah salah satu operasi paling penting dalam linear algebra.

Untuk dua vektor:
a = [a1, a2, ..., an]
b = [b1, b2, ..., bn]

dot product didefinisikan sebagai:
a · b = a1b1 + a2b2 + ... + anbn

Jadi:
1. kalikan elemen yang posisinya sama
2. jumlahkan seluruh hasil perkalian

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

dot_manual = np.sum(a * b)
dot_numpy = np.dot(a, b)

print("a =", a)
print("b =", b)
print("Dot product manual =", dot_manual)
print("Dot product NumPy =", dot_numpy)
```

## 13. Memahami Dot Product Secara Langkah demi Langkah

Kita lihat proses perhitungannya:
- elemen pertama: 1 × 4
- elemen kedua: 2 × 5
- elemen ketiga: 3 × 6

Kemudian semua hasil dijumlahkan.

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

elementwise_product = a * b
dot_result = np.sum(elementwise_product)

step_df = pd.DataFrame({
    "a": a,
    "b": b,
    "a*b": elementwise_product
})

print("Dot product =", dot_result)
step_df
```

## 14. Makna Geometris Dot Product

Dot product juga punya makna geometris:

a · b = ||a|| ||b|| cos(theta)

di mana:
- ||a|| = panjang vektor a
- ||b|| = panjang vektor b
- theta = sudut antara a dan b

Artinya:
- jika dua vektor searah, dot product besar dan positif
- jika saling tegak lurus, dot product = 0
- jika berlawanan arah, dot product bisa bernilai negatif

## 15. Menghitung Sudut Antar Vektor dari Dot Product

Dari rumus:
a · b = ||a|| ||b|| cos(theta)

maka:
cos(theta) = (a · b) / (||a|| ||b||)

dan:
theta = arccos(cos(theta))

```python
a = np.array([3, 1])
b = np.array([1, 3])

dot_ab = np.dot(a, b)
norm_a = np.linalg.norm(a)
norm_b = np.linalg.norm(b)

cos_theta = dot_ab / (norm_a * norm_b)
theta_rad = np.arccos(cos_theta)
theta_deg = np.degrees(theta_rad)

print("a =", a)
print("b =", b)
print("Dot product =", round(dot_ab, 4))
print("Norm a =", round(norm_a, 4))
print("Norm b =", round(norm_b, 4))
print("cos(theta) =", round(cos_theta, 4))
print("Sudut (radian) =", round(theta_rad, 4))
print("Sudut (derajat) =", round(theta_deg, 4))
```

## 16. Contoh Kasus Vektor Tegak Lurus (Orthogonal)

Dua vektor dikatakan **orthogonal** jika sudut di antara keduanya adalah 90 derajat.

Ciri penting:
- jika dua vektor orthogonal, dot product-nya sama dengan 0

```python
u = np.array([1, 2])
t = np.array([2, -1])

dot_ut = np.dot(u, t)

print("u =", u)
print("t =", t)
print("u · t =", dot_ut)
```

Jika dua vektor orthogonal, secara geometris mereka saling tegak lurus.

```python
u = np.array([1, 2])
t = np.array([2, -1])

plt.figure(figsize=(7, 7))
plt.axhline(0, color="gray", linewidth=1)
plt.axvline(0, color="gray", linewidth=1)

plt.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, label='u')
plt.quiver(0, 0, t[0], t[1], angles='xy', scale_units='xy', scale=1, label='t')

plt.xlim(-3, 4)
plt.ylim(-3, 4)
plt.xlabel("Sumbu X")
plt.ylabel("Sumbu Y")
plt.title("Visualisasi Dua Vektor Orthogonal")
plt.legend()
plt.grid(True)
plt.show()
```

## 19. Proyeksi Vektor

Salah satu konsep penting dari dot product adalah **proyeksi**.

Proyeksi vektor `a` ke arah vektor `b` memberi tahu:
- seberapa besar komponen `a` yang berada pada arah `b`

Rumus proyeksi vektor `a` pada `b`:

proj_b(a) = ((a · b) / (b · b)) * b

Konsep ini sangat penting pada:
- dekomposisi sinyal
- reduksi dimensi
- PCA
- least squares
- regresi linear

```python
a = np.array([4, 2])
b = np.array([1, 3])

projection_of_a_on_b = (np.dot(a, b) / np.dot(b, b)) * b

print("a =", a)
print("b =", b)
print("Proyeksi a pada b =", projection_of_a_on_b)
```

## 20. Visualisasi Proyeksi Vektor

Pada visualisasi berikut:
- `a` adalah vektor asli
- `b` adalah arah acuan
- `proj_b(a)` adalah bagian dari `a` yang berada pada arah `b`

```python
a = np.array([4, 2])
b = np.array([1, 3])
proj = (np.dot(a, b) / np.dot(b, b)) * b

plt.figure(figsize=(7, 7))
plt.axhline(0, color="gray", linewidth=1)
plt.axvline(0, color="gray", linewidth=1)

plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, label='a')
plt.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, label='b')
plt.quiver(0, 0, proj[0], proj[1], angles='xy', scale_units='xy', scale=1, label='proj_b(a)')

plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.xlabel("Sumbu X")
plt.ylabel("Sumbu Y")
plt.title("Visualisasi Proyeksi Vektor")
plt.legend()
plt.grid(True)
plt.show()
```

## 21. Memisahkan Vektor Menjadi Komponen Sejajar dan Tegak Lurus

Sebuah vektor `a` dapat dipecah menjadi dua bagian:
1. bagian yang sejajar dengan `b`
2. bagian yang tegak lurus terhadap `b`

Secara umum:
a = proj_b(a) + komponen_orthogonal

dengan:
komponen_orthogonal = a - proj_b(a)

```python
a = np.array([4, 2])
b = np.array([1, 3])

proj = (np.dot(a, b) / np.dot(b, b)) * b
orth_component = a - proj

print("a =", a)
print("proj_b(a) =", proj)
print("Komponen orthogonal =", orth_component)

print("\nCek bahwa a = proj + orth_component")
print(proj + orth_component)
```

## 22. Verifikasi Bahwa Komponen Orthogonal Benar-Benar Tegak Lurus

Jika `orth_component` benar-benar orthogonal terhadap `b`, maka:
orth_component · b = 0
atau sangat dekat dengan 0 karena pembulatan numerik.

```python
dot_check = np.dot(orth_component, b)
print("Dot product komponen orthogonal dengan b =", dot_check)
```

## 23. Contoh Dot Product sebagai Similarity Sederhana

Dot product sering digunakan untuk mengukur kemiripan arah antar-vektor.

Sebagai ilustrasi, misalkan kita punya dua siswa dengan skor pada tiga mata pelajaran:
- Matematika
- Fisika
- Pemrograman

Semakin mirip pola nilainya, biasanya dot product akan semakin besar jika besar vektornya juga sebanding.

```python
student_A = np.array([80, 75, 90])
student_B = np.array([78, 74, 92])
student_C = np.array([30, 95, 20])

dot_AB = np.dot(student_A, student_B)
dot_AC = np.dot(student_A, student_C)

similarity_df = pd.DataFrame({
    "Perbandingan": ["A vs B", "A vs C"],
    "Dot Product": [dot_AB, dot_AC]
})

similarity_df
```

## 24. Catatan Penting tentang Dot Product sebagai Similarity

Walaupun dot product bisa memberi gambaran kemiripan, ada satu hal penting:
- dot product dipengaruhi oleh **besar vektor**
- jadi dua vektor yang besar bisa menghasilkan dot product besar walaupun pola arahnya tidak sepenuhnya identik

Karena itu, untuk similarity yang lebih adil, sering digunakan **cosine similarity**, yaitu dot product yang dinormalisasi oleh norm masing-masing vektor.

## 25. Cosine Similarity

Rumus cosine similarity:

cos(theta) = (a · b) / (||a|| ||b||)

Nilainya berada di sekitar:
- mendekati 1 → sangat mirip arah
- mendekati 0 → tidak mirip / orthogonal
- mendekati -1 → berlawanan arah

```python
def cosine_similarity(x, y):
    return np.dot(x, y) / (np.linalg.norm(x) * np.linalg.norm(y))

cos_AB = cosine_similarity(student_A, student_B)
cos_AC = cosine_similarity(student_A, student_C)

cos_df = pd.DataFrame({
    "Perbandingan": ["A vs B", "A vs C"],
    "Cosine Similarity": [cos_AB, cos_AC]
})

cos_df
```
