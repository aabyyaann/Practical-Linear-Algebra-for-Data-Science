## 4. Kelemahan dan Tantangan Utama Proyek

Walaupun proyek ini memiliki banyak sisi positif, terdapat beberapa kelemahan dan tantangan yang perlu diperhatikan.

### a. Integrasi IoT berpotensi kompleks
Salah satu tantangan utama adalah proses integrasi perangkat IoT dengan backend dan dashboard web. Komunikasi data, format payload, kestabilan koneksi, dan sinkronisasi data real-time dapat menjadi hambatan teknis.

### b. Risiko penambahan ruang lingkup
Jika di tengah proyek muncul permintaan fitur baru, timeline dan biaya dapat terganggu. Ini dikenal sebagai **scope creep**.

### c. Keterbatasan sumber daya tim
Jika jumlah anggota tim sedikit, maka ada kemungkinan satu orang harus menangani banyak pekerjaan sekaligus, seperti coding, testing, dokumentasi, dan deployment. Hal ini dapat menurunkan efisiensi.

Dalam banyak proyek mahasiswa, pengujian sering baru dilakukan di akhir. Padahal jika bug ditemukan terlambat, maka waktu perbaikan akan semakin sempit.

### e. Ketergantungan pada perangkat dan layanan tertentu

## 5. Rekomendasi Perbaikan Proyek

Berdasarkan hasil analisis, berikut beberapa rekomendasi yang dapat diterapkan agar proyek berjalan lebih baik:

### 1. Menetapkan prioritas fitur sejak awal
Fitur inti harus dikerjakan terlebih dahulu, misalnya:
- monitoring data sensor,
- dashboard utama,
- penyimpanan data,
- tampilan status perangkat.

Fitur tambahan seperti analitik lanjutan, notifikasi cerdas, atau integrasi tambahan dapat dikerjakan setelah fitur utama stabil.

### 2. Melakukan testing secara bertahap
Setiap modul yang selesai sebaiknya langsung diuji, misalnya:
- uji API backend,
- uji tampilan frontend,
- uji koneksi perangkat IoT,
- uji database.

Dengan cara ini, bug dapat ditemukan lebih cepat dan tidak menumpuk di akhir.

### 3. Menambahkan buffer waktu pada timeline
Fase integrasi dan pengujian sebaiknya memiliki cadangan waktu karena dua fase ini paling sering mengalami hambatan.

### 4. Menetapkan penanggung jawab tiap pekerjaan
Setiap aktivitas utama perlu memiliki PIC (Person in Charge) agar tanggung jawab lebih jelas dan tidak saling menunggu.

### 5. Menyediakan contingency budget
Walaupun skala proyek akademik, tetap lebih baik jika ada cadangan biaya untuk kebutuhan tak terduga seperti revisi perangkat, hosting, atau pembelian komponen tambahan.

### 6. Membuat dokumentasi progres secara berkala
Setiap perkembangan proyek, perubahan fitur, kendala, dan keputusan teknis sebaiknya didokumentasikan. Hal ini akan membantu saat evaluasi maupun penyusunan laporan akhir.
