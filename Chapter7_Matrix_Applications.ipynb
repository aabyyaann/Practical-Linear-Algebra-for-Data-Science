# Dasar Teori

## 1. Matriks sebagai Transformasi Linear
Jika kita memiliki vektor \(x\) dan matriks \(A\), maka hasil:

\[
y = Ax
\]

dapat dipandang sebagai **transformasi** dari vektor \(x\) menjadi vektor baru \(y\).

Transformasi ini dapat berupa:
- scaling (pembesaran / pengecilan)
- rotasi
- refleksi
- shear
- kombinasi beberapa transformasi

---

## 2. Matriks sebagai Operator
Sebuah matriks bukan hanya kumpulan angka, tetapi dapat dianggap sebagai **fungsi linear** yang menerima input vektor dan menghasilkan output vektor baru.

Jika:
\[
A \in \mathbb{R}^{m \times n}
\]
dan
\[
x \in \mathbb{R}^{n}
\]

maka:
\[
Ax \in \mathbb{R}^{m}
\]

Jadi matriks memetakan vektor dari satu ruang ke ruang lain.

---

## 3. Komposisi Transformasi
Jika ada dua transformasi:
- transformasi pertama oleh matriks \(A\)
- transformasi kedua oleh matriks \(B\)

maka komposisi transformasi dapat ditulis sebagai:

\[
B(Ax) = (BA)x
\]

Artinya, hasil gabungan transformasi dapat direpresentasikan sebagai perkalian matriks.

---

## 4. Sistem Persamaan Linear
Banyak masalah aljabar linear dapat ditulis sebagai:

\[
Ax = b
\]

di mana:
- \(A\) = matriks koefisien
- \(x\) = vektor variabel yang ingin dicari
- \(b\) = vektor hasil

Bentuk ini sangat penting dalam:
- regresi
- optimisasi
- simulasi numerik
- machine learning
- pemrosesan sinyal

---

## 5. Matriks dalam Representasi Data
Pada data science, dataset sering direpresentasikan sebagai matriks:
- baris = sampel / observasi
- kolom = fitur / variabel

Contoh:
\[
X =
\begin{bmatrix}
x_{11} & x_{12} & \dots & x_{1p} \\
x_{21} & x_{22} & \dots & x_{2p} \\
\vdots & \vdots & \ddots & \vdots \\
x_{n1} & x_{n2} & \dots & x_{np}
\end{bmatrix}
\]

Representasi ini menjadi dasar hampir semua algoritma machine learning.

---

## 6. Geometri Matriks
Secara geometris, matriks dapat:
- mengubah panjang vektor
- mengubah arah vektor
- mengubah bentuk himpunan titik
- memetakan persegi menjadi jajaran genjang
- memetakan lingkaran menjadi elips

Karena itu, matriks sangat penting untuk memahami transformasi ruang.

```python
import numpy as np
import matplotlib.pyplot as plt

np.set_printoptions(precision=3, suppress=True)
```

# 1. Matriks sebagai Transformasi pada Sebuah Vektor

Kita mulai dari contoh paling dasar: sebuah matriks mengubah satu vektor menjadi vektor baru.

```python
A = np.array([
    [2, 0],
    [0, 1]
])

x = np.array([1, 2])
y = A @ x

print("Matriks A:")
print(A)

print("\nVektor x:")
print(x)

print("\nHasil transformasi y = A @ x:")
print(y)
```

# Analisis Transformasi Vektor

Matriks:

\[
A =
\begin{bmatrix}
2 & 0 \\
0 & 1
\end{bmatrix}
\]

diterapkan ke vektor:

\[
x =
\begin{bmatrix}
1 \\
2
\end{bmatrix}
\]

menghasilkan:

\[
y =
\begin{bmatrix}
2 \\
2
\end{bmatrix}
\]

Interpretasinya:
- komponen pertama dikali 2
- komponen kedua tetap

Jadi transformasi ini adalah **scaling pada sumbu x**.

# 2. Visualisasi Vektor Sebelum dan Sesudah Transformasi

```python
origin = np.array([0, 0])

plt.figure(figsize=(6, 6))
plt.quiver(*origin, x[0], x[1], angles='xy', scale_units='xy', scale=1, label='Vektor x')
plt.quiver(*origin, y[0], y[1], angles='xy', scale_units='xy', scale=1, label='Vektor y = Ax')

plt.xlim(0, 4)
plt.ylim(0, 4)
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.grid(True)
plt.legend()
plt.title("Transformasi Vektor oleh Matriks A")
plt.show()
```

# Analisis Visualisasi

Grafik memperlihatkan:
- vektor asli \(x = [1,2]\)
- vektor hasil transformasi \(y = [2,2]\)

Terlihat bahwa vektor hasil lebih “melebar” ke arah sumbu x.  
Inilah salah satu cara termudah memahami bahwa matriks bekerja sebagai operator transformasi.

# 3. Transformasi Scaling pada Banyak Titik

Sekarang kita tidak hanya mentransformasi satu vektor, tetapi beberapa titik sekaligus.

```python
points = np.array([
    [0, 0],
    [1, 0],
    [1, 1],
    [0, 1]
])

A = np.array([
    [2, 0],
    [0, 1]
])

transformed_points = (A @ points.T).T

print("Titik awal:")
print(points)

print("\nTitik setelah transformasi:")
print(transformed_points)
```

# Analisis Sekumpulan Titik

Kita menggunakan empat titik yang membentuk persegi satuan:

\[
(0,0), (1,0), (1,1), (0,1)
\]

Setelah dikalikan dengan matriks:

\[
A =
\begin{bmatrix}
2 & 0 \\
0 & 1
\end{bmatrix}
\]

koordinat x menjadi dua kali lebih besar, sedangkan koordinat y tetap.  
Akibatnya, persegi berubah menjadi persegi panjang yang lebih lebar.

# 4. Visualisasi Transformasi Sekumpulan Titik

```python
plt.figure(figsize=(7, 6))

# titik asli
plt.scatter(points[:, 0], points[:, 1], label='Titik asli')
for i, p in enumerate(points):
    plt.text(p[0] + 0.03, p[1] + 0.03, f"P{i}")

# titik hasil transformasi
plt.scatter(transformed_points[:, 0], transformed_points[:, 1], label='Titik transformasi')
for i, p in enumerate(transformed_points):
    plt.text(p[0] + 0.03, p[1] + 0.03, f"T{i}")

plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.grid(True)
plt.legend()
plt.title("Transformasi Sekumpulan Titik oleh Matriks")
plt.show()
```

# Analisis Geometris

Visualisasi ini menunjukkan bagaimana matriks dapat mengubah bentuk geometri.

Sebelum transformasi:
- titik membentuk persegi

Setelah transformasi:
- bentuk melebar di arah horizontal

Aplikasi ide ini sangat luas, misalnya:
- grafika komputer
- augmentasi data
- transformasi koordinat
- pemrosesan citra

# 5. Transformasi Rotasi 2D

Salah satu aplikasi matriks yang sangat penting adalah **rotasi**.
Matriks rotasi 2D dengan sudut \(\theta\) adalah:

\[
R =
\begin{bmatrix}
\cos \theta & -\sin \theta \\
\sin \theta & \cos \theta
\end{bmatrix}
\]

```python
theta = np.deg2rad(45)

R = np.array([
    [np.cos(theta), -np.sin(theta)],
    [np.sin(theta),  np.cos(theta)]
])

v = np.array([2, 1])
v_rot = R @ v

print("Matriks rotasi R:")
print(R)

print("\nVektor awal:", v)
print("Vektor setelah rotasi 45 derajat:", v_rot)
```

# Analisis Rotasi

Matriks rotasi memutar vektor tanpa mengubah strukturnya secara linear.  
Jika sudut rotasi adalah 45°, maka vektor akan berpindah arah berlawanan arah jarum jam sebesar 45°.

Rotasi sangat penting dalam:
- grafika komputer
- robotika
- computer vision
- transformasi koordinat

# 6. Visualisasi Rotasi Vektor

```python
plt.figure(figsize=(6, 6))
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, label='Vektor awal')
plt.quiver(0, 0, v_rot[0], v_rot[1], angles='xy', scale_units='xy', scale=1, label='Vektor setelah rotasi')

plt.xlim(-1, 3)
plt.ylim(-1, 3)
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.grid(True)
plt.legend()
plt.title("Rotasi Vektor 2D")
plt.show()
```

# Analisis Visualisasi Rotasi

Grafik menunjukkan perubahan arah vektor setelah dikalikan dengan matriks rotasi.  
Panjang vektor relatif tetap, tetapi orientasinya berubah.

Hal ini menunjukkan bahwa matriks tidak selalu mengubah panjang; matriks juga bisa hanya mengubah arah.

# 7. Refleksi dengan Matriks

Refleksi terhadap sumbu-x dapat ditulis sebagai:

\[
F =
\begin{bmatrix}
1 & 0 \\
0 & -1
\end{bmatrix}
\]

Refleksi ini mempertahankan koordinat x dan membalik tanda koordinat y.

```python
F = np.array([
    [1, 0],
    [0, -1]
])

v = np.array([2, 3])
v_reflect = F @ v

print("Matriks refleksi:")
print(F)

print("\nVektor awal:", v)
print("Vektor hasil refleksi:", v_reflect)
```

# Analisis Refleksi

Jika vektor awal adalah:

\[
[2,3]
\]

maka setelah refleksi terhadap sumbu-x menjadi:

\[
[2,-3]
\]

Jadi:
- nilai x tetap
- nilai y berubah tanda

# 8. Komposisi Transformasi

Sekarang kita gabungkan dua transformasi:
1. scaling
2. rotasi

Jika vektor \(x\) terlebih dahulu dikalikan matriks \(A\), lalu hasilnya dikalikan matriks \(B\), maka:

\[
y = B(Ax) = (BA)x
\]

Artinya, dua transformasi dapat digabung menjadi satu matriks gabungan.

```python
S = np.array([
    [2, 0],
    [0, 1]
])

theta = np.deg2rad(30)
R = np.array([
    [np.cos(theta), -np.sin(theta)],
    [np.sin(theta),  np.cos(theta)]
])

v = np.array([1, 1])

step1 = S @ v
step2 = R @ step1

combined = R @ S
combined_result = combined @ v

print("Hasil scaling lalu rotasi:", step2)
print("Hasil matriks gabungan:", combined_result)
```

# Analisis Komposisi Transformasi

Kita melakukan dua cara:
1. transformasi bertahap: scaling lalu rotasi
2. transformasi langsung dengan matriks gabungan \(RS\)

Hasil keduanya sama. Ini membuktikan bahwa komposisi transformasi linear dapat direpresentasikan dengan perkalian matriks.

# 9. Matriks sebagai Representasi Sistem Persamaan Linear

Misalkan kita punya sistem:

\[
2x + y = 5
\]
\[
x + 3y = 7
\]

Sistem ini dapat ditulis sebagai:

\[
Ax = b
\]

dengan:
\[
A =
\begin{bmatrix}
2 & 1 \\
1 & 3
\end{bmatrix},
\quad
x =
\begin{bmatrix}
x \\
y
\end{bmatrix},
\quad
b =
\begin{bmatrix}
5 \\
7
\end{bmatrix}
\]

```python
A = np.array([
    [2, 1],
    [1, 3]
])

b = np.array([5, 7])

solution = np.linalg.solve(A, b)

print("Matriks koefisien A:")
print(A)

print("\nVektor b:")
print(b)

print("\nSolusi [x, y]:")
print(solution)
```

# Analisis Sistem Linear

Hasil `np.linalg.solve(A, b)` memberikan solusi sistem persamaan linear.

Jika hasilnya misalnya:
\[
x = 1.6,\quad y = 1.8
\]

maka nilai tersebut memenuhi kedua persamaan sekaligus.

Bentuk matriks seperti ini sangat penting karena banyak masalah machine learning dapat ditulis dalam bentuk:

\[
Ax = b
\]

atau variasi yang lebih umum.

# 10. Verifikasi Solusi Sistem Linear

```python
x_sol = solution[0]
y_sol = solution[1]

eq1 = 2*x_sol + y_sol
eq2 = x_sol + 3*y_sol

print("2x + y =", eq1)
print("x + 3y =", eq2)
```

# Analisis Verifikasi

Nilai yang diperoleh harus mendekati:
- persamaan pertama = 5
- persamaan kedua = 7

Jika benar, maka solusi matriks sudah sesuai.

# 15. Matriks untuk Transformasi Banyak Vektor Sekaligus

Salah satu kekuatan matriks adalah kemampuannya mentransformasi banyak vektor sekaligus.

Kita buat sekumpulan titik 2D lalu transformasikan semuanya dalam satu operasi.

```python
pts = np.array([
    [1, 0],
    [0, 1],
    [1, 1],
    [2, 1],
    [1, 2]
])

T = np.array([
    [1, 1],
    [0, 1]
])

pts_transformed = (T @ pts.T).T

print("Titik awal:")
print(pts)

print("\nTitik setelah transformasi:")
print(pts_transformed)
```

# Analisis Transformasi Banyak Titik

Di sini setiap titik dianggap sebagai vektor 2D.  
Semua titik ditransformasikan oleh matriks yang sama.

Ini sangat penting dalam komputasi modern karena:
- lebih efisien
- lebih ringkas
- sesuai dengan struktur data berbasis array / tensor

# 16. Visualisasi Sebelum dan Sesudah Transformasi Banyak Titik

```python
plt.figure(figsize=(7, 6))

plt.scatter(pts[:, 0], pts[:, 1], label='Titik awal')
for i, p in enumerate(pts):
    plt.text(p[0] + 0.03, p[1] + 0.03, f"P{i}")

plt.scatter(pts_transformed[:, 0], pts_transformed[:, 1], label='Titik transformasi')
for i, p in enumerate(pts_transformed):
    plt.text(p[0] + 0.03, p[1] + 0.03, f"T{i}")

plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.grid(True)
plt.legend()
plt.title("Transformasi Banyak Titik oleh Matriks")
plt.show()
```

# Analisis Visualisasi Transformasi Banyak Titik

Dari visualisasi terlihat bahwa seluruh pola titik berubah mengikuti aturan matriks transformasi.

Hal ini menunjukkan bahwa matriks dapat digunakan untuk:
- memodifikasi sekumpulan data sekaligus
- memetakan koordinat
- mengubah struktur geometri objek

# 17. Mengukur Luas Transformasi secara Intuitif dengan Determinan

Determinan berkaitan dengan perubahan skala area/volume oleh transformasi linear.

Jika sebuah matriks 2×2 memiliki determinan:
- besar absolut > 1 → area membesar
- antara 0 dan 1 → area mengecil
- negatif → orientasi juga terbalik
- 0 → ruang “runtuh” ke dimensi lebih rendah

```python
A1 = np.array([
    [2, 0],
    [0, 1]
])

A2 = np.array([
    [1, 2],
    [2, 4]
])

print("det(A1) =", np.linalg.det(A1))
print("det(A2) =", np.linalg.det(A2))
```

# Analisis Determinan sebagai Ukuran Skala Transformasi

## Matriks A1
Determinan A1 tidak nol, sehingga transformasinya masih mempertahankan dimensi ruang.

## Matriks A2
Determinan A2 = 0, sehingga transformasi ini meruntuhkan ruang ke dimensi lebih rendah.  
Akibatnya, matriks tidak memiliki invers dan informasi hilang.

# 18. Aplikasi Sederhana ke Machine Learning: Prediksi Linear

Salah satu bentuk paling dasar model machine learning adalah:

\[
y = Xw
\]

di mana:
- \(X\) = matriks fitur
- \(w\) = vektor bobot
- \(y\) = prediksi

Kita akan membuat contoh kecil untuk melihat ide ini.

```python
X = np.array([
    [1, 2],
    [2, 1],
    [3, 4]
])

w = np.array([0.5, 1.0])

y_pred = X @ w

print("Matriks fitur X:")
print(X)

print("\nBobot w:")
print(w)

print("\nPrediksi linear y = X @ w:")
print(y_pred)
```

# Analisis Model Linear

Setiap prediksi dihitung sebagai kombinasi linear dari fitur-fitur pada satu baris data.

Sebagai contoh untuk sampel pertama:

\[
y_1 = 1(0.5) + 2(1.0) = 2.5
\]

Konsep sederhana ini adalah fondasi dari:
- regresi linear
- logistic regression (dengan fungsi tambahan)
- perceptron
- neural network layer
