# Dasar Teori

## 1. Perkalian Matriks
Jika:
- matriks \(A\) berukuran \(m \times n\)
- matriks \(B\) berukuran \(n \times p\)

maka hasil perkalian:

\[
C = AB
\]

akan berukuran:

\[
m \times p
\]

Elemen hasil perkalian diperoleh dari **dot product** antara baris matriks pertama dan kolom matriks kedua.

---

## 2. Syarat Perkalian Matriks
Dua matriks dapat dikalikan jika:
- jumlah kolom matriks pertama = jumlah baris matriks kedua

Misalnya:
- \(A\) berukuran \(2 \times 3\)
- \(B\) berukuran \(3 \times 4\)

maka \(AB\) valid dan hasilnya berukuran \(2 \times 4\).

---

## 3. Perkalian Matriks ≠ Perkalian Elemen per Elemen
Ada dua operasi yang berbeda:
1. **perkalian matriks** → menggunakan aturan aljabar linear
2. **perkalian elemen per elemen** → setiap elemen dikalikan dengan elemen pada posisi yang sama

Di NumPy:
- `A @ B` atau `np.matmul(A,B)` → perkalian matriks
- `A * B` → perkalian elemen per elemen

---

## 4. Diagonal Matriks
Elemen diagonal utama matriks adalah elemen:
\[
a_{11}, a_{22}, a_{33}, \dots
\]

Diagonal matriks penting dalam banyak operasi, misalnya:
- trace
- matriks diagonal
- eigenvalue
- kovarians

---

## 5. Trace Matriks
Trace matriks adalah jumlah elemen-elemen diagonal utama:

\[
\text{tr}(A) = \sum_i a_{ii}
\]

Trace hanya didefinisikan untuk **matriks persegi**.

---

## 6. Determinan
Determinan adalah nilai skalar yang dihitung dari matriks persegi.  
Determinan memberikan informasi penting, misalnya:
- apakah matriks memiliki invers
- apakah transformasi linear “meruntuhkan” dimensi
- apakah matriks singular atau tidak

Jika determinan = 0, maka matriks **tidak memiliki invers**.

---

## 7. Rank Matriks
Rank matriks adalah jumlah maksimum baris/kolom yang saling independen secara linear.

Rank membantu kita memahami:
- banyaknya informasi unik dalam matriks
- apakah ada redundansi antar baris/kolom
- apakah sistem persamaan memiliki solusi unik atau tidak

---

## 8. Matriks Simetris
Matriks simetris memenuhi:

\[
A = A^T
\]

Artinya, elemen di atas diagonal sama dengan elemen di bawah diagonal.

Matriks simetris sangat penting pada:
- matriks kovarians
- quadratic forms
- optimisasi
- PCA

```python
import numpy as np
import matplotlib.pyplot as plt

np.set_printoptions(precision=3, suppress=True)
```

# 1. Membuat Dua Matriks untuk Operasi Lanjutan

Kita akan mulai dengan membuat dua matriks yang kompatibel untuk dikalikan.

```python
A = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

B = np.array([
    [7, 8],
    [9, 10],
    [11, 12]
])

print("Matriks A:")
print(A)
print("Shape A:", A.shape)

print("\nMatriks B:")
print(B)
print("Shape B:", B.shape)
```

# Analisis Matriks A dan B

Matriks A berukuran:

\[
2 \times 3
\]

Matriks B berukuran:

\[
3 \times 2
\]

Karena jumlah kolom A = jumlah baris B = 3, maka perkalian matriks \(AB\) **valid**.

Ukuran hasil perkalian nanti adalah:

\[
2 \times 2
\]

# 2. Perkalian Matriks

```python
C = A @ B

print("Hasil A @ B:")
print(C)
print("Shape hasil:", C.shape)
```

C = A @ B

# 3. Verifikasi Perkalian Matriks Secara Manual

Untuk memastikan konsepnya benar, kita hitung satu elemen hasil secara manual.

```python
manual_c11 = A[0, 0]*B[0, 0] + A[0, 1]*B[1, 0] + A[0, 2]*B[2, 0]
manual_c12 = A[0, 0]*B[0, 1] + A[0, 1]*B[1, 1] + A[0, 2]*B[2, 1]

print("c11 manual =", manual_c11)
print("c12 manual =", manual_c12)
print("Baris pertama hasil perkalian =", C[0])
```

# Analisis Verifikasi

Hasil perhitungan manual harus sama dengan elemen pada hasil `A @ B`.  
Ini menunjukkan bahwa perkalian matriks bukan sekadar mengalikan angka satu per satu, melainkan menjumlahkan hasil kali antara:
- satu baris
- dan satu kolom

Inilah inti dari operasi matriks dalam aljabar linear.

# 4. Perkalian Elemen per Elemen

Sekarang kita bandingkan dengan perkalian **elemen per elemen**.  
Agar bisa dilakukan, kedua matriks harus memiliki shape yang sama.

```python
X = np.array([
    [1, 2],
    [3, 4]
])

Y = np.array([
    [10, 20],
    [30, 40]
])

elemwise = X * Y

print("Matriks X:")
print(X)

print("\nMatriks Y:")
print(Y)

print("\nPerkalian elemen per elemen X * Y:")
print(elemwise)
```

# Analisis Perkalian Elemen per Elemen

Pada operasi `X * Y`, setiap elemen dikalikan dengan elemen pada posisi yang sama:

\[
\begin{bmatrix}
1 & 2 \\
3 & 4
\end{bmatrix}
*
\begin{bmatrix}
10 & 20 \\
30 & 40
\end{bmatrix}
=
\begin{bmatrix}
10 & 40 \\
90 & 160
\end{bmatrix}
\]

Ini **berbeda** dengan perkalian matriks.  
Di machine learning, operasi elemen per elemen sering muncul pada:
- masking
- aktivasi neural network
- pembobotan fitur tertentu

# 5. Perbandingan Perkalian Matriks vs Elemen per Elemen

```python
M1 = np.array([
    [1, 2],
    [3, 4]
])

M2 = np.array([
    [5, 6],
    [7, 8]
])

print("M1:")
print(M1)

print("\nM2:")
print(M2)

print("\nM1 @ M2 (perkalian matriks):")
print(M1 @ M2)

print("\nM1 * M2 (perkalian elemen per elemen):")
print(M1 * M2)
```

# Analisis Perbedaan Dua Operasi

Untuk matriks yang sama, hasil:
- `M1 @ M2`
- `M1 * M2`

akan berbeda karena definisinya berbeda.

## Perkalian matriks
Menggunakan kombinasi baris × kolom.

## Perkalian elemen per elemen
Menggunakan pasangan elemen pada posisi yang sama.

Membedakan kedua operasi ini sangat penting agar tidak salah memahami implementasi NumPy maupun algoritma machine learning.

# 6. Diagonal Matriks

Sekarang kita lihat bagaimana mengambil diagonal utama dari sebuah matriks persegi.

```python
D = np.array([
    [2, 1, 4],
    [0, 3, 5],
    [7, 8, 6]
])

diag_D = np.diag(D)

print("Matriks D:")
print(D)

print("\nDiagonal utama D:")
print(diag_D)
```

# Analisis Diagonal Matriks

Diagonal utama matriks D adalah elemen:
- \(d_{11}\)
- \(d_{22}\)
- \(d_{33}\)

Pada contoh ini, diagonalnya adalah:

\[
[2, 3, 6]
\]

Diagonal matriks sering digunakan untuk:
- menghitung trace

# Analisis Trace

Trace adalah jumlah elemen diagonal utama:

\[
\text{tr}(D) = 2 + 3 + 6 = 11
\]

Trace hanya didefinisikan untuk matriks persegi.  
Dalam aljabar linear dan statistik, trace sering muncul pada:
- analisis kovarians

# 8. Matriks Simetris

Matriks disebut simetris jika:

\[
A = A^T
\]

Artinya, matriks sama dengan transposenya.

```python
S = np.array([
    [1, 2, 3],
    [2, 5, 6],
    [3, 6, 9]
])

print("Matriks S:")
print(S)

print("\nTranspose S:")
print(S.T)

print("\nApakah S simetris?", np.array_equal(S, S.T))
```

# Analisis Matriks Simetris

Matriks simetris memiliki pola:
- elemen di atas diagonal = elemen di bawah diagonal

Pada matriks S:
- \(s_{12} = s_{21}\)
- \(s_{13} = s_{31}\)
- \(s_{23} = s_{32}\)

Matriks simetris sangat penting dalam machine learning, misalnya pada:
- matriks kovarians
- matriks kernel tertentu
- optimisasi kuadratik

# 9. Determinan Matriks

Determinan hanya didefinisikan untuk matriks persegi.  
Determinan membantu kita mengetahui apakah matriks:
- dapat dibalik (invertible)
- singular
- atau memiliki transformasi yang “meruntuhkan” ruang

```python
M = np.array([
    [2, 1],
    [4, 3]
])

det_M = np.linalg.det(M)

print("Matriks M:")
print(M)
print("\nDeterminan M =", det_M)
```

# Analisis Determinan

Untuk matriks 2×2:

\[
M =
\begin{bmatrix}
a & b \\
c & d
\end{bmatrix}
\]

determinan dihitung dengan:

\[
\det(M) = ad - bc
\]

Pada contoh:

\[
\det(M) = (2)(3) - (1)(4) = 6 - 4 = 2
\]

Karena determinannya **tidak nol**, maka matriks M:
- tidak singular
- memiliki invers

# 10. Contoh Matriks Singular

Sekarang kita buat matriks yang determinannya nol.

```python
singular_matrix = np.array([
    [1, 2],
    [2, 4]
])

det_sing = np.linalg.det(singular_matrix)

print("Matriks singular:")
print(singular_matrix)
print("\nDeterminan =", det_sing)
```

# Analisis Matriks Singular

Matriks:

\[
\begin{bmatrix}
1 & 2 \\
2 & 4
\end{bmatrix}
\]

memiliki baris kedua = 2 × baris pertama.  
Artinya, baris-barisnya **tidak independen secara linear**.

Akibatnya:
- determinan = 0
- matriks singular
- matriks tidak memiliki invers

# 11. Rank Matriks

Rank matriks menyatakan jumlah maksimum baris atau kolom yang saling independen secara linear.

```python
R1 = np.array([
    [1, 2],
    [3, 4]
])

R2 = np.array([
    [1, 2],
    [2, 4]
])

print("Matriks R1:")
print(R1)
print("Rank R1 =", np.linalg.matrix_rank(R1))

print("\nMatriks R2:")
print(R2)
print("Rank R2 =", np.linalg.matrix_rank(R2))
```

# Analisis Rank

## Matriks R1
Matriks R1 memiliki rank 2 karena kedua barisnya independen secara linear.

## Matriks R2
Matriks R2 memiliki rank 1 karena baris kedua adalah kelipatan dari baris pertama.

Rank sangat penting karena menunjukkan:
- seberapa banyak informasi unik dalam matriks
- apakah ada redundansi
- apakah sistem linear dapat memiliki solusi unik

# 12. Rank pada Matriks yang Lebih Besar

```python
R3 = np.array([
    [1, 2, 3],
    [2, 4, 6],
    [1, 1, 1]
])

print("Matriks R3:")
print(R3)
print("Rank R3 =", np.linalg.matrix_rank(R3))
```

# Analisis Rank Matriks R3

Pada matriks R3:
- baris kedua = 2 × baris pertama
- sehingga tidak memberikan informasi baru

Namun baris ketiga berbeda dan independen terhadap baris pertama.  
Akibatnya, rank matriks ini adalah **2**, bukan 3.

# 13. Matriks Diagonal

Matriks diagonal adalah matriks persegi yang hanya memiliki nilai nonnol pada diagonal utama.

```python
diag_values = np.array([5, 10, 15])
Diag = np.diag(diag_values)

print("Nilai diagonal:")
print(diag_values)

print("\nMatriks diagonal:")
print(Diag)
```

# Analisis Matriks Diagonal

Jika diberikan vektor:

\[
[5,10,15]
\]

maka matriks diagonalnya adalah:

\[
\begin{bmatrix}
5 & 0 & 0 \\
0 & 10 & 0 \\
0 & 0 & 15
\end{bmatrix}
\]

Matriks diagonal penting karena:
- perhitungannya relatif sederhana
- sering muncul pada eigenvalue
- berguna dalam skala transformasi per sumbu

# 14. Matriks sebagai Transformasi Linear 2D

Salah satu makna penting matriks adalah sebagai **transformasi linear**.  
Misalnya, matriks dapat mengubah posisi vektor di bidang 2D.

```python
v = np.array([1, 1])

T = np.array([
    [2, 0],
    [0, 1]
])

v_transformed = T @ v

print("Vektor asli:", v)
print("Matriks transformasi T:")
print(T)
print("Vektor hasil transformasi:", v_transformed)
```

# Analisis Transformasi Linear

Matriks transformasi:

\[
T =
\begin{bmatrix}
2 & 0 \\
0 & 1
\end{bmatrix}
\]

mengubah vektor:

\[
v =
\begin{bmatrix}
1 \\
1
\end{bmatrix}
\]

menjadi:

\[
Tv =
\begin{bmatrix}
2 \\
1
\end{bmatrix}
\]

Artinya:
- komponen x dikali 2
- komponen y tetap

Transformasi ini dapat dipandang sebagai **stretching** pada sumbu x.

# 15. Visualisasi Transformasi Linear Sederhana

```python
origin = np.array([0, 0])

plt.figure(figsize=(6, 6))
plt.quiver(*origin, v[0], v[1], angles='xy', scale_units='xy', scale=1, label='Vektor asli')
plt.quiver(*origin, v_transformed[0], v_transformed[1], angles='xy', scale_units='xy', scale=1, label='Vektor transformasi')

plt.xlim(0, 3)
plt.ylim(0, 3)
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.grid(True)
plt.legend()
plt.title("Visualisasi Transformasi Linear")
plt.show()
```

# Analisis Visualisasi

Grafik menunjukkan dua vektor:
- **vektor asli** \([1,1]\)
- **vektor hasil transformasi** \([2,1]\)

Dari visualisasi ini terlihat bahwa matriks dapat mengubah:
- panjang vektor
- arah vektor
- skala ruang

Konsep transformasi linear ini nantinya sangat penting pada:
- rotasi
- refleksi
- scaling
- PCA
- perubahan basis

# 16. Sifat Dasar Perkalian Matriks terhadap Identitas

Matriks identitas adalah “angka 1”-nya matriks.
Jika ukuran cocok, maka:

\[
AI = IA = A
\]

```python
P = np.array([
    [3, 1],
    [2, 4]
])

I = np.eye(2)

print("Matriks P:")
print(P)

print("\nP @ I:")
print(P @ I)

print("\nI @ P:")
print(I @ P)
```

# Analisis Matriks Identitas

Matriks identitas tidak mengubah matriks lain saat dikalikan, selama ukurannya sesuai.

Ini sama seperti pada bilangan real:
\[
a \times 1 = a
\]

Dalam aljabar linear:
\[
AI = IA = A
\]
