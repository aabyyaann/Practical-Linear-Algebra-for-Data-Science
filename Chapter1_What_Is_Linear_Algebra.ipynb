## 2. Apa Itu Linear Algebra?

Linear algebra adalah cabang matematika yang mempelajari objek seperti:
- **skalar**: satu nilai tunggal
- **vektor**: kumpulan nilai yang memiliki arah/komponen
- **matriks**: susunan angka dalam baris dan kolom
- **tensor**: generalisasi dari matriks ke dimensi yang lebih tinggi

Dalam data science, linear algebra sangat penting karena hampir semua data numerik dapat direpresentasikan dalam bentuk vektor atau matriks.

Contohnya:
- dataset tabular direpresentasikan sebagai matriks
- parameter model machine learning sering disimpan sebagai vektor
- transformasi data, PCA, regresi, neural network, dan optimisasi semuanya menggunakan operasi linear algebra

## 3. Skalar, Vektor, Matriks, dan Tensor

Sebelum masuk ke algoritma yang lebih kompleks, kita perlu memahami empat objek dasar:

### Skalar
Skalar adalah satu nilai tunggal, misalnya:
- 5
- -2.7
- 100

### Vektor
Vektor adalah sekumpulan angka yang disusun dalam satu dimensi.  
Contoh:
- [1, 2, 3]
- [10, 20, 30, 40]

### Matriks
Matriks adalah susunan angka dalam bentuk baris dan kolom.  
Contoh matriks 2×3:
[
 [1, 2, 3],
 [4, 5, 6]
]

### Tensor
Tensor adalah generalisasi array ke dimensi yang lebih tinggi.  
Contohnya:
- vektor = tensor 1D
- matriks = tensor 2D
- kumpulan matriks = tensor 3D

```python
# Contoh skalar
scalar = 7

# Contoh vektor
vector = np.array([1, 2, 3, 4])

# Contoh matriks
matrix = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

# Contoh tensor 3D
tensor = np.array([
    [[1, 2], [3, 4]],
    [[5, 6], [7, 8]]
])

print("Skalar:")
print(scalar)

print("\nVektor:")
print(vector)

print("\nMatriks:")
print(matrix)

print("\nTensor 3D:")
print(tensor)
```

## 4. Melihat Bentuk (Shape) dan Dimensi Data

Dalam linear algebra dan data science, sangat penting untuk memahami **shape** dari data.

- **Shape** menunjukkan ukuran array pada tiap dimensi
- **ndim** menunjukkan jumlah dimensi
- **size** menunjukkan total jumlah elemen

Konsep ini sangat penting karena banyak operasi linear algebra hanya valid jika shape antar-objek sesuai.

```python
objects_info = pd.DataFrame({
    "Objek": ["Scalar", "Vector", "Matrix", "Tensor"],
    "Nilai": [
        scalar,
        str(vector),
        str(matrix),
        str(tensor)
    ],
    "Shape": [
        np.array(scalar).shape,
        vector.shape,
        matrix.shape,
        tensor.shape
    ],
    "Jumlah Dimensi": [
        np.array(scalar).ndim,
        vector.ndim,
        matrix.ndim,
        tensor.ndim
    ],
    "Jumlah Elemen": [
        np.array(scalar).size,
        vector.size,
        matrix.size,
        tensor.size
    ]
})

objects_info
```

## 5. Mengapa Linear Algebra Penting dalam Data Science?

Linear algebra menjadi fondasi data science karena data dan model biasanya direpresentasikan dalam bentuk numerik terstruktur.

Beberapa contoh penerapannya:
1. **Dataset sebagai matriks**
   - setiap baris mewakili observasi
   - setiap kolom mewakili fitur

2. **Model machine learning**
   - bobot model sering berbentuk vektor atau matriks

3. **Transformasi data**
   - normalisasi, rotasi, proyeksi, dan reduksi dimensi menggunakan operasi matriks

4. **Optimisasi**
   - gradien, turunan parsial, dan update parameter memanfaatkan operasi vektor/matriks

5. **Deep learning**
   - input, bobot, aktivasi, dan output di neural network semuanya berbasis linear algebra

## 6. Dataset sebagai Matriks

Dalam machine learning, dataset umumnya direpresentasikan sebagai matriks **X**:
- baris = sampel / observasi
- kolom = fitur / variabel

Sebagai contoh, misalkan kita memiliki data 5 mahasiswa dengan 3 fitur:
- nilai matematika
- nilai fisika
- nilai pemrograman

Maka dataset tersebut dapat ditulis sebagai matriks 5×3.

```python
students_scores = np.array([
    [80, 75, 90],
    [60, 70, 65],
    [85, 88, 92],
    [70, 68, 72],
    [95, 90, 98]
])

df_scores = pd.DataFrame(
    students_scores,
    columns=["Matematika", "Fisika", "Pemrograman"],
    index=[f"Mahasiswa_{i}" for i in range(1, 6)]
)

df_scores
```

## 7. Interpretasi Matriks Dataset

Pada tabel sebelumnya:
- setiap **baris** merepresentasikan satu mahasiswa
- setiap **kolom** merepresentasikan satu fitur/nilai
- seluruh tabel dapat dianggap sebagai sebuah **matriks data**

Bentuk matriks:
- jumlah baris = 5
- jumlah kolom = 3

Dalam notasi linear algebra, jika ada **m** sampel dan **n** fitur, maka dataset dapat ditulis sebagai matriks berukuran **m × n**.

```python
print("Shape dataset:", students_scores.shape)
print("Jumlah sampel (baris):", students_scores.shape[0])
print("Jumlah fitur (kolom):", students_scores.shape[1])
```

## 8. Operasi Dasar pada Vektor

Vektor mendukung berbagai operasi dasar, misalnya:
- penjumlahan
- pengurangan
- perkalian dengan skalar
- rata-rata
- jumlah total

Operasi-operasi ini sangat sering digunakan dalam preprocessing data, optimisasi, dan evaluasi model.

```python
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])

print("v1 =", v1)
print("v2 =", v2)

print("\nPenjumlahan v1 + v2 =", v1 + v2)
print("Pengurangan v1 - v2 =", v1 - v2)
print("Perkalian skalar 3 * v1 =", 3 * v1)
print("Jumlah elemen v1 =", np.sum(v1))
print("Rata-rata elemen v1 =", np.mean(v1))
```

## 9. Operasi Dasar pada Matriks

Matriks juga mendukung operasi dasar, seperti:
- penjumlahan matriks
- pengurangan matriks
- perkalian elemen per elemen
- transpose
- pencarian nilai maksimum, minimum, rata-rata, dan sebagainya

Pada chapter awal ini kita fokus dulu pada pemahaman bentuk dan operasi sederhana.

```python
A = np.array([
    [1, 2],
    [3, 4]
])

B = np.array([
    [5, 6],
    [7, 8]
])

print("Matrix A:")
print(A)

print("\nMatrix B:")
print(B)

print("\nA + B:")
print(A + B)

print("\nA - B:")
print(A - B)

print("\nPerkalian elemen-per-elemen A * B:")
print(A * B)

print("\nTranspose A:")
print(A.T)
```

## 10. Visualisasi Matriks sebagai Tabel Angka

Matriks sering lebih mudah dipahami jika ditampilkan dalam bentuk tabel atau heatmap.  
Visualisasi ini membantu kita melihat pola angka, perbedaan nilai, dan struktur data.

## 11. Vektor sebagai Representasi Satu Observasi

Dalam machine learning, satu observasi/sampel dapat direpresentasikan sebagai vektor.

Misalnya satu mahasiswa memiliki tiga fitur:
- nilai matematika
- nilai fisika
- nilai pemrograman

Maka satu mahasiswa dapat direpresentasikan sebagai vektor 3 dimensi.

```python
student_1 = students_scores[0]

print("Vektor mahasiswa pertama:")
print(student_1)

print("\nJumlah fitur:", len(student_1))
```

## 12. Kolom sebagai Vektor Fitur

Tidak hanya baris, kolom pada dataset juga dapat dipandang sebagai vektor.

Contohnya:
- kolom nilai Matematika adalah satu vektor
- kolom nilai Fisika adalah satu vektor
- kolom nilai Pemrograman adalah satu vektor

Ini penting karena banyak analisis statistik dan machine learning bekerja pada hubungan antar fitur.

```python
math_scores = students_scores[:, 0]
physics_scores = students_scores[:, 1]
programming_scores = students_scores[:, 2]

feature_vectors = pd.DataFrame({
    "Matematika": math_scores,
    "Fisika": physics_scores,
    "Pemrograman": programming_scores
})

feature_vectors
```

## 13. Rata-rata Fitur sebagai Ringkasan Data

Salah satu operasi paling umum pada matriks data adalah menghitung rata-rata tiap kolom.

Secara praktis:
- rata-rata tiap kolom menunjukkan pusat data untuk masing-masing fitur
- ini sering digunakan pada statistik deskriptif, normalisasi, dan standardisasi

```python
mean_per_feature = students_scores.mean(axis=0)

summary_df = pd.DataFrame({
    "Fitur": ["Matematika", "Fisika", "Pemrograman"],
    "Rata-rata": mean_per_feature
})

summary_df
```

## 15. Hubungan Linear Algebra dan Machine Learning

Linear algebra bukan sekadar teori matematika, tetapi merupakan dasar komputasi untuk berbagai algoritma machine learning.

Contohnya:
- **Linear Regression** menggunakan kombinasi linear antar fitur
- **Logistic Regression** memanfaatkan operasi vektor dan matriks
- **PCA** menggunakan eigenvalue/eigenvector
- **SVD** digunakan untuk reduksi dimensi dan rekomendasi sistem
- **Neural Network** melakukan perkalian matriks pada setiap layer

## 16. Contoh Sederhana Prediksi Linear

Salah satu ide paling dasar dalam machine learning adalah menggabungkan beberapa fitur dengan bobot tertentu untuk menghasilkan sebuah skor prediksi.

Misalnya:
- fitur = nilai matematika, fisika, pemrograman
- bobot = seberapa penting masing-masing fitur

Prediksi sederhana dapat ditulis sebagai kombinasi linear:
prediksi = w1*x1 + w2*x2 + w3*x3

Di linear algebra, ini adalah operasi antara vektor fitur dan vektor bobot.

```python
# Misal bobot untuk tiga fitur
weights = np.array([0.3, 0.2, 0.5])

# Prediksi skor untuk mahasiswa pertama
prediction_student_1 = np.sum(student_1 * weights)

print("Fitur mahasiswa pertama:", student_1)
print("Bobot:", weights)
print("Skor prediksi linear:", round(prediction_student_1, 2))
```

## 17. Prediksi untuk Semua Sampel

Karena seluruh dataset disimpan sebagai matriks, kita dapat menghitung skor untuk semua mahasiswa sekaligus menggunakan operasi array.

Inilah salah satu kekuatan linear algebra:
- komputasi menjadi ringkas
- lebih cepat
- mudah diskalakan ke data yang lebih besar

```python
predictions = students_scores @ weights

prediction_df = df_scores.copy()
prediction_df["Skor_Prediksi"] = predictions
prediction_df
```

## 18. Interpretasi Operasi Matrix-Vector

Pada cell sebelumnya, kita menggunakan operator `@` untuk menghitung:

**students_scores @ weights**

Artinya:
- `students_scores` adalah matriks berukuran 5×3
- `weights` adalah vektor berukuran 3
- hasilnya adalah vektor berukuran 5

Setiap nilai output mewakili skor prediksi satu mahasiswa.  
Konsep ini sangat penting karena menjadi dasar dari banyak model machine learning.

```python
print("Shape matriks data:", students_scores.shape)
print("Shape vektor bobot:", weights.shape)
print("Shape hasil prediksi:", predictions.shape)
```

## 19. Membandingkan Nilai Asli dan Skor Prediksi

Walaupun contoh ini sangat sederhana, kita bisa melihat bagaimana sebuah dataset dapat diubah menjadi skor numerik menggunakan kombinasi linear.

Pada machine learning sesungguhnya:
- bobot tidak ditentukan manual
- bobot dipelajari dari data
- fungsi loss digunakan untuk menilai kesalahan prediksi
- optimisasi digunakan untuk memperbaiki bobot

```python
plt.figure(figsize=(8, 5))
plt.plot(predictions, marker="o")
plt.title("Skor Prediksi Linear Tiap Mahasiswa")
plt.xlabel("Index Mahasiswa")
plt.ylabel("Skor Prediksi")
plt.show()
```
