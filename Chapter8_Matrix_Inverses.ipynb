## Ringkasan Teori

### 1. Apa itu invers matriks?
Untuk sebuah matriks persegi \( A \), invers matriks ditulis sebagai \( A^{-1} \), yaitu matriks yang memenuhi:

\[
A A^{-1} = A^{-1} A = I
\]

dengan \( I \) adalah **identity matrix**.

Artinya, invers matriks berperan seperti kebalikan pada bilangan biasa:
- jika \( a \neq 0 \), maka \( a \cdot \frac{1}{a} = 1 \)
- jika \( A \) invertible, maka \( A A^{-1} = I \)

---

### 2. Kapan matriks punya invers?
Suatu matriks **hanya dapat memiliki invers jika**:
- matriksnya **persegi** (\( n \times n \))
- **determinannya tidak nol**
- rank penuh (*full rank*)
- kolom/barisnya saling independen linear

Jika determinan = 0, maka matriks disebut **singular** dan **tidak punya invers**.

---

### 3. Mengapa invers penting?
Jika kita punya sistem linear:

\[
A\mathbf{x} = \mathbf{b}
\]

dan \( A \) invertible, maka solusi bisa ditulis:

\[
\mathbf{x} = A^{-1}\mathbf{b}
\]

Walaupun secara teori benar, dalam praktik komputasi kita **lebih disarankan menggunakan** `np.linalg.solve(A,b)` daripada menghitung invers secara eksplisit, karena:
- lebih stabil secara numerik
- lebih efisien
- lebih cepat

---

### 4. Hubungan dengan identitas
Jika \( A^{-1} \) benar, maka:

\[
A A^{-1} = I
\]

dan hasilnya harus menjadi identity matrix.  
Karena komputer bekerja dengan bilangan floating point, hasil numerik kadang tidak persis 0 atau 1, tetapi sangat dekat.

---

### 5. Intuisi geometris
Jika sebuah matriks melakukan transformasi linear terhadap ruang, maka invers matriks adalah transformasi yang **membalikkan efek tersebut**.

Contoh:
- jika \( A \) meregangkan dan memutar vektor,
- maka \( A^{-1} \) akan mengembalikan vektor ke posisi semula.

````python
# Import library utama
import numpy as np
import matplotlib.pyplot as plt

np.set_printoptions(precision=4, suppress=True)
````

## 1. Membuat matriks persegi

Kita mulai dengan sebuah matriks 2x2 sederhana yang memiliki invers.

````python
A = np.array([
    [2, 1],
    [1, 1]
], dtype=float)

print("Matriks A:")
print(A)
print("Shape A:", A.shape)
````

## 2. Menghitung invers matriks dengan `np.linalg.inv()`

NumPy menyediakan fungsi `np.linalg.inv()` untuk menghitung invers matriks.

````python
A_inv = np.linalg.inv(A)

print("Invers dari A:")
print(A_inv)
````

## 3. Verifikasi sifat invers

Jika invers benar, maka:

\[
A A^{-1} = I
\]

dan juga:

\[
A^{-1} A = I
\]

Kita cek keduanya.

````python
left_product = A @ A_inv
right_product = A_inv @ A

print("A @ A_inv =")
print(left_product)

print("\nA_inv @ A =")
print(right_product)
````

## 4. Membandingkan dengan identity matrix

Hasil perkalian seharusnya sama dengan matriks identitas.

````python
I = np.eye(A.shape[0])

print("Identity matrix I:")
print(I)

print("\nApakah A @ A_inv mendekati I?")
print(np.allclose(left_product, I))

print("\nApakah A_inv @ A mendekati I?")
print(np.allclose(right_product, I))
````

## Penjelasan
Fungsi `np.allclose()` digunakan karena hasil komputasi floating point sering tidak persis sama, tetapi sangat dekat.  
Jika hasil `True`, maka secara numerik perkalian tersebut dianggap sama dengan matriks identitas.

## 5. Contoh matriks 3x3

Sekarang kita coba matriks yang lebih besar.

````python
B = np.array([
    [1, 2, 3],
    [0, 1, 4],
    [5, 6, 0]
], dtype=float)

print("Matriks B:")
print(B)
````

````python
B_inv = np.linalg.inv(B)

print("Invers matriks B:")
print(B_inv)
````

````python
print("B @ B_inv =")
print(B @ B_inv)

print("\nApakah mendekati identity?")
print(np.allclose(B @ B_inv, np.eye(3)))
````

## 6. Determinan dan keberadaan invers

Salah satu cara paling umum untuk mengecek apakah matriks punya invers adalah melihat determinannya.

- Jika `det(A) != 0` → matriks invertible
- Jika `det(A) == 0` → matriks singular, tidak punya invers

````python
det_A = np.linalg.det(A)
det_B = np.linalg.det(B)

print("det(A) =", det_A)
print("det(B) =", det_B)
````

## 7. Contoh matriks singular

Sekarang kita buat matriks yang **tidak punya invers**.

````python
C = np.array([
    [1, 2],
    [2, 4]
], dtype=float)

print("Matriks C:")
print(C)

print("\nDeterminan C:")
print(np.linalg.det(C))
````

Perhatikan bahwa baris kedua merupakan kelipatan dari baris pertama.  
Artinya, matriks ini **tidak full rank** dan determinannya 0, sehingga matriks tersebut singular.

````python
try:
    C_inv = np.linalg.inv(C)
    print(C_inv)
except np.linalg.LinAlgError as e:
    print("Error:", e)
````

## Penjelasan
Error di atas menunjukkan bahwa matriks singular tidak dapat diinverskan.  
Ini penting dalam machine learning dan komputasi numerik, karena matriks yang singular sering menyebabkan solusi tidak unik atau bahkan tidak bisa dihitung dengan metode tertentu.

# 8. Menggunakan invers untuk menyelesaikan sistem linear

Misalkan kita punya sistem:

\[
A\mathbf{x} = \mathbf{b}
\]

Jika \( A \) invertible, maka:

\[
\mathbf{x} = A^{-1}\mathbf{b}
\]

Mari kita coba dengan contoh sederhana.

````python
A = np.array([
    [3, 1],
    [1, 2]
], dtype=float)

b = np.array([9, 8], dtype=float)

print("A =")
print(A)

print("\nb =")
print(b)
````

````python
A_inv = np.linalg.inv(A)
x_via_inverse = A_inv @ b

print("Solusi x menggunakan invers:")
print(x_via_inverse)
````

## 9. Verifikasi solusi

Kita cek apakah hasil \( x \) benar-benar memenuhi \( A x = b \).

````python
check_b = A @ x_via_inverse

print("A @ x =")
print(check_b)

print("\nApakah sama dengan b?")
print(np.allclose(check_b, b))
````

# 10. Cara yang lebih baik: `np.linalg.solve()`

Walaupun rumus teorinya adalah:

\[
x = A^{-1}b
\]

dalam praktik komputasi, cara yang lebih baik adalah:

```python
np.linalg.solve(A, b)

````python
x_via_solve = np.linalg.solve(A, b)

print("Solusi x menggunakan np.linalg.solve():")
print(x_via_solve)
````

````python
print("Perbandingan hasil:")
print("x via inverse =", x_via_inverse)
print("x via solve   =", x_via_solve)

print("\nApakah keduanya sama?")
print(np.allclose(x_via_inverse, x_via_solve))
````

## Penjelasan penting
Secara matematis, kedua pendekatan memberi hasil yang sama.  
Namun untuk aplikasi nyata, terutama matriks besar, **`np.linalg.solve()` lebih direkomendasikan** daripada menghitung invers lalu mengalikan dengan \( b \).

# 11. Visualisasi ide “transformasi lalu dibalik”

Secara geometris, matriks dapat dipandang sebagai transformasi linear.  
Jika sebuah vektor \( x \) ditransformasikan oleh matriks \( A \), maka:

\[
y = Ax
\]

Kemudian jika kita kalikan dengan inversnya:

\[
x = A^{-1}y
\]

vektor akan kembali ke posisi semula.

````python
A = np.array([
    [2, 1],
    [1, 1]
], dtype=float)

A_inv = np.linalg.inv(A)

x = np.array([1, 2], dtype=float)
y = A @ x
x_recovered = A_inv @ y

print("x asli        =", x)
print("y = A @ x     =", y)
print("x dipulihkan  =", x_recovered)
````

````python
A = np.array([
    [2, 1],
    [1, 1]
], dtype=float)

A_inv = np.linalg.inv(A)

x = np.array([1, 2], dtype=float)
y = A @ x
x_recovered = A_inv @ y

print("x asli        =", x)
print("y = A @ x     =", y)
print("x dipulihkan  =", x_recovered)
````

## Interpretasi visual
- **x asli** adalah vektor awal
- **y = A @ x** adalah hasil transformasi oleh matriks \( A \)
- **A_inv @ y** mengembalikan vektor ke posisi semula

Ini menunjukkan makna geometris dari invers: **membatalkan efek transformasi linear**.

````python
D = np.diag([2, 4, 5])

print("Matriks diagonal D:")
print(D)

D_inv = np.linalg.inv(D)

print("\nInvers D:")
print(D_inv)

print("\nVerifikasi D @ D_inv:")
print(D @ D_inv)
````

# 13. Hubungan invers dengan rank

Matriks yang invertible harus memiliki:
- rank penuh
- determinan tidak nol
- kolom independen linear
- baris independen linear

Mari kita bandingkan matriks invertible dan singular.

````python
M1 = np.array([
    [1, 2],
    [3, 4]
], dtype=float)

M2 = np.array([
    [1, 2],
    [2, 4]
], dtype=float)

print("Rank M1 =", np.linalg.matrix_rank(M1))
print("det(M1) =", np.linalg.det(M1))

print("\nRank M2 =", np.linalg.matrix_rank(M2))
print("det(M2) =", np.linalg.det(M2))
````

## Interpretasi
- `M1` memiliki rank penuh (2), sehingga invertible
- `M2` hanya rank 1, sehingga singular dan tidak punya invers

Ini memperlihatkan hubungan erat antara:
- rank
- determinan
- independensi linear
- keberadaan invers

# 14. Contoh numerik yang sedikit lebih realistis

Kita buat sistem linear 3 variabel dan selesaikan dengan dua cara:
1. menggunakan invers
2. menggunakan `np.linalg.solve()`

````python
A = np.array([
    [2, 1, -1],
    [-3, -1, 2],
    [-2, 1, 2]
], dtype=float)

b = np.array([8, -11, -3], dtype=float)

print("A =")
print(A)
print("\nb =", b)
````

````python
x_inv = np.linalg.inv(A) @ b
x_solve = np.linalg.solve(A, b)

print("x dengan invers:")
print(x_inv)

print("\nx dengan solve:")
print(x_solve)
````

````python
print("Verifikasi A @ x_inv:")
print(A @ x_inv)

print("\nVerifikasi A @ x_solve:")
print(A @ x_solve)
````

# 15. Kelemahan menghitung invers secara eksplisit

Walaupun secara teori mudah dipahami, menghitung invers secara eksplisit punya beberapa kelemahan:

1. **lebih mahal secara komputasi**
2. **kurang stabil secara numerik**
3. bisa memperbesar error floating point
4. tidak diperlukan jika tujuan hanya menyelesaikan sistem linear

Karena itu:
- untuk teori → invers sangat penting
- untuk implementasi numerik → `solve()` biasanya lebih baik

# 16. Mini eksperimen: matriks acak invertible

Sekarang kita buat beberapa matriks acak dan memeriksa apakah:
- determinannya tidak nol
- hasil \( A A^{-1} \) mendekati identitas

````python
np.random.seed(42)

for i in range(3):
    M = np.random.randn(3, 3)

    # pastikan tidak singular secara numerik
    if abs(np.linalg.det(M)) < 1e-8:
        print(f"Matriks ke-{i+1} terlalu dekat singular, dilewati")
        continue

    M_inv = np.linalg.inv(M)
    product = M @ M_inv

    print(f"\n=== Matriks ke-{i+1} ===")
    print("det(M) =", np.linalg.det(M))
    print("Apakah M @ M_inv mendekati I?", np.allclose(product, np.eye(3)))
````
