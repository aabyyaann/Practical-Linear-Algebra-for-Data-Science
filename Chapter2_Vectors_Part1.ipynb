## 2. Apa Itu Vektor?

Vektor adalah kumpulan angka yang disusun dalam satu dimensi.  
Dalam konteks linear algebra, vektor biasanya digunakan untuk merepresentasikan:
- satu observasi data,
- satu fitur dari dataset,
- bobot model machine learning,
- arah dan besar suatu perubahan,
- atau sekumpulan parameter numerik.

Contoh sederhana vektor:
- [1, 2, 3]
- [10, 20, 30, 40]
- [-2, 5, 8]

Dalam data science, vektor sangat penting karena hampir semua data numerik pada akhirnya diproses sebagai vektor atau kumpulan vektor.

## 3. Membuat Vektor dengan NumPy

Di Python, vektor paling mudah direpresentasikan menggunakan `numpy.array()`.

Secara teknis, NumPy menyimpan vektor sebagai array 1 dimensi.  
Contoh:
- vektor dengan 3 elemen
- vektor dengan 5 elemen
- vektor berisi bilangan pecahan

```python
v = np.array([1, 2, 3])
w = np.array([10, 20, 30, 40, 50])
u = np.array([1.5, -2.0, 3.25])

print("Vektor v:", v)
print("Vektor w:", w)
print("Vektor u:", u)
```

## 4. Shape, Dimensi, dan Panjang Vektor

Tiga informasi penting dari sebuah vektor adalah:

1. **Shape**
   - menunjukkan ukuran array
   - untuk vektor 1D dengan 3 elemen, shape-nya adalah `(3,)`

2. **Jumlah dimensi (`ndim`)**
   - menunjukkan banyaknya sumbu pada array
   - vektor biasa NumPy memiliki dimensi 1

3. **Panjang vektor**
   - jumlah elemen dalam vektor

```python
vector_info = pd.DataFrame({
    "Nama Vektor": ["v", "w", "u"],
    "Isi": [str(v), str(w), str(u)],
    "Shape": [v.shape, w.shape, u.shape],
    "Jumlah Dimensi": [v.ndim, w.ndim, u.ndim],
    "Jumlah Elemen": [len(v), len(w), len(u)]
})

vector_info
```

## 5. Mengakses Elemen Vektor

Setiap elemen pada vektor memiliki posisi atau indeks.  
Di Python, indeks dimulai dari **0**.

Contoh:
- elemen pertama = indeks 0
- elemen kedua = indeks 1
- elemen terakhir bisa diakses dengan indeks `-1`

```python
print("Vektor v:", v)
print("Elemen pertama v:", v[0])
print("Elemen kedua v:", v[1])
print("Elemen ketiga v:", v[2])
print("Elemen terakhir v:", v[-1])
```

## 6. Row Vector dan Column Vector

Dalam linear algebra, vektor bisa ditulis dalam dua bentuk umum:

### Row vector
Ditulis secara horizontal:
[1, 2, 3]

### Column vector
Ditulis secara vertikal:
[[1],
 [2],
 [3]]

Di NumPy:
- array 1D seperti `[1,2,3]` belum secara eksplisit row atau column
- untuk membuat column vector, kita perlu mengubah shape menjadi `(n,1)`
- untuk row vector 2D, shape-nya biasanya `(1,n)`

```python
v_1d = np.array([1, 2, 3])
v_row = np.array([[1, 2, 3]])
v_col = np.array([[1], [2], [3]])

print("Vektor 1D:")
print(v_1d)
print("shape:", v_1d.shape)

print("\nRow vector:")
print(v_row)
print("shape:", v_row.shape)

print("\nColumn vector:")
print(v_col)
print("shape:", v_col.shape)
```

## 7. Mengubah Vektor 1D Menjadi Row/Column Vector

Dalam praktik data science, kita sering perlu mengubah bentuk vektor:
- dari 1D ke row vector
- dari 1D ke column vector

Hal ini penting karena beberapa operasi linear algebra memerlukan shape tertentu agar bisa dijalankan.

```python
x = np.array([4, 5, 6, 7])

x_row = x.reshape(1, -1)   # 1 baris, jumlah kolom otomatis
x_col = x.reshape(-1, 1)   # jumlah baris otomatis, 1 kolom

print("Vektor asli:", x)
print("Shape asli:", x.shape)

print("\nRow vector:")
print(x_row)
print("Shape row:", x_row.shape)

print("\nColumn vector:")
print(x_col)
print("Shape column:", x_col.shape)
```

## 8. Vektor Nol, Vektor Satu, dan Vektor Rentang

NumPy menyediakan cara cepat untuk membuat beberapa jenis vektor umum:

- `np.zeros(n)` → vektor berisi nol
- `np.ones(n)` → vektor berisi satu
- `np.arange()` → vektor dengan nilai berurutan
- `np.linspace()` → vektor dengan interval teratur

```python
zeros_vec = np.zeros(5)
ones_vec = np.ones(5)
range_vec = np.arange(1, 6)
linspace_vec = np.linspace(0, 1, 6)

print("Vektor nol:", zeros_vec)
print("Vektor satu:", ones_vec)
print("Vektor rentang (arange):", range_vec)
print("Vektor linspace:", linspace_vec)
```

## 9. Operasi Dasar pada Vektor

Operasi dasar yang umum dilakukan pada vektor meliputi:
- penjumlahan
- pengurangan
- perkalian dengan skalar
- pembagian dengan skalar
- operasi elemen per elemen

Operasi-operasi ini penting karena menjadi dasar komputasi pada model machine learning.

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print("a =", a)
print("b =", b)

print("\na + b =", a + b)
print("a - b =", a - b)
print("2 * a =", 2 * a)
print("a / 2 =", a / 2)
print("a * b =", a * b)   # perkalian elemen per elemen
```

## 10. Menjumlahkan dan Merata-ratakan Elemen Vektor

Selain operasi antar-vektor, kita juga sering ingin menghitung ringkasan statistik sederhana dari sebuah vektor:
- jumlah total elemen
- rata-rata
- nilai maksimum
- nilai minimum

```python
sample_vec = np.array([3, 6, 9, 12, 15])

summary = pd.DataFrame({
    "Statistik": ["Jumlah", "Rata-rata", "Minimum", "Maksimum"],
    "Nilai": [
        np.sum(sample_vec),
        np.mean(sample_vec),
        np.min(sample_vec),
        np.max(sample_vec)
    ]
})

summary
```

## 11. Vektor sebagai Satu Observasi Data

Dalam data science, satu observasi dapat direpresentasikan sebagai vektor.

Misalnya sebuah data rumah memiliki 4 fitur:
- luas bangunan
- jumlah kamar
- jumlah kamar mandi
- harga listrik bulanan

Maka satu rumah dapat direpresentasikan sebagai satu vektor dengan 4 elemen.

```python
house_1 = np.array([120, 3, 2, 450000])

print("Representasi satu rumah sebagai vektor:")
print(house_1)
print("Jumlah fitur:", len(house_1))
```

## 12. Sekumpulan Observasi sebagai Matriks

Jika satu observasi adalah vektor, maka sekumpulan observasi dapat disusun menjadi matriks.

Contoh:
- 5 rumah
- masing-masing rumah memiliki 4 fitur

Maka dataset tersebut dapat direpresentasikan sebagai matriks 5×4.

```python
houses = np.array([
    [120, 3, 2, 450000],
    [90,  2, 1, 300000],
    [150, 4, 3, 600000],
    [110, 3, 2, 400000],
    [80,  2, 1, 250000]
])

houses_df = pd.DataFrame(
    houses,
    columns=["Luas", "Kamar", "Kamar_Mandi", "Listrik"],
    index=[f"Rumah_{i}" for i in range(1, 6)]
)

houses_df
```

## 13. Baris dan Kolom sebagai Vektor

Dari matriks dataset:
- setiap **baris** adalah vektor observasi
- setiap **kolom** adalah vektor fitur

Keduanya penting:
- vektor baris digunakan untuk melihat satu sampel
- vektor kolom digunakan untuk menganalisis satu fitur pada semua sampel

```python
rumah_pertama = houses[0, :]
fitur_luas = houses[:, 0]

print("Vektor observasi rumah pertama:")
print(rumah_pertama)

print("\nVektor fitur luas semua rumah:")
print(fitur_luas)
```

## 14. Visualisasi Vektor pada Bidang 2D

Jika sebuah vektor hanya memiliki dua elemen, kita bisa memvisualisasikannya sebagai panah pada bidang 2D.

Contoh:
vektor [3, 2] berarti:
- bergerak 3 satuan pada sumbu x
- bergerak 2 satuan pada sumbu y

```python
vec2d = np.array([3, 2])

plt.figure(figsize=(6, 6))
plt.axhline(0, color="gray", linewidth=1)
plt.axvline(0, color="gray", linewidth=1)

plt.quiver(0, 0, vec2d[0], vec2d[1], angles='xy', scale_units='xy', scale=1)
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.title("Visualisasi Vektor 2D")
plt.xlabel("Sumbu X")
plt.ylabel("Sumbu Y")
plt.grid(True)
plt.show()
```

## 15. Beberapa Vektor pada Bidang 2D

Agar lebih memahami vektor sebagai arah dan besar, kita dapat menggambar beberapa vektor sekaligus pada bidang yang sama.

```python
vectors = {
    "v1": np.array([2, 1]),
    "v2": np.array([1, 3]),
    "v3": np.array([4, 2])
}

plt.figure(figsize=(7, 7))
plt.axhline(0, color="gray", linewidth=1)
plt.axvline(0, color="gray", linewidth=1)

for name, vec in vectors.items():
    plt.quiver(0, 0, vec[0], vec[1], angles='xy', scale_units='xy', scale=1, label=name)

plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.xlabel("Sumbu X")
plt.ylabel("Sumbu Y")
plt.title("Beberapa Vektor 2D")
plt.legend()
plt.grid(True)
plt.show()
```

## 16. Panjang Vektor Secara Geometris

Sebuah vektor memiliki besar atau panjang.  
Secara intuitif, panjang vektor menunjukkan seberapa jauh panah tersebut dari titik asal `(0,0)`.

Untuk vektor 2D `[x, y]`, panjang vektor dapat dipahami melalui teorema Pythagoras:
panjang = sqrt(x² + y²)

Di NumPy, kita bisa menghitungnya dengan `np.linalg.norm()`.

```python
vec_length = np.array([3, 4])

length_manual = np.sqrt(vec_length[0]**2 + vec_length[1]**2)
length_numpy = np.linalg.norm(vec_length)

print("Vektor:", vec_length)
print("Panjang vektor (manual):", length_manual)
print("Panjang vektor (NumPy):", length_numpy)
```

## 17. Vektor Negatif

Jika semua arah suatu vektor dibalik, kita memperoleh vektor negatif.

Misalnya:
- v = [2, 3]
- -v = [-2, -3]

Secara geometris, vektor negatif menunjuk ke arah berlawanan tetapi memiliki panjang yang sama.

```python
v_pos = np.array([2, 3])
v_neg = -v_pos

plt.figure(figsize=(7, 7))
plt.axhline(0, color="gray", linewidth=1)
plt.axvline(0, color="gray", linewidth=1)

plt.quiver(0, 0, v_pos[0], v_pos[1], angles='xy', scale_units='xy', scale=1, label="v")
plt.quiver(0, 0, v_neg[0], v_neg[1], angles='xy', scale_units='xy', scale=1, label="-v")

plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.xlabel("Sumbu X")
plt.ylabel("Sumbu Y")
plt.title("Vektor dan Vektor Negatif")
plt.legend()
plt.grid(True)
plt.show()
```

## 18. Perkalian Skalar terhadap Vektor

Ketika sebuah vektor dikalikan dengan skalar:
- jika skalar > 1, vektor menjadi lebih panjang
- jika 0 < skalar < 1, vektor menjadi lebih pendek
- jika skalar negatif, arah vektor dibalik

Operasi ini penting dalam transformasi data dan optimisasi.

```python
base_vec = np.array([2, 1])
scaled_1 = 2 * base_vec
scaled_2 = 0.5 * base_vec
scaled_3 = -1 * base_vec

vectors_scaled = {
    "base": base_vec,
    "2 * base": scaled_1,
    "0.5 * base": scaled_2,
    "-1 * base": scaled_3
}

plt.figure(figsize=(8, 8))
plt.axhline(0, color="gray", linewidth=1)
plt.axvline(0, color="gray", linewidth=1)

for name, vec in vectors_scaled.items():
    plt.quiver(0, 0, vec[0], vec[1], angles='xy', scale_units='xy', scale=1, label=name)

plt.xlim(-3, 5)
plt.ylim(-3, 5)
plt.xlabel("Sumbu X")
plt.ylabel("Sumbu Y")
plt.title("Pengaruh Perkalian Skalar terhadap Vektor")
plt.legend()
plt.grid(True)
plt.show()
```

## 19. Menyusun Tabel Panjang Beberapa Vektor

Agar lebih rapi, kita bisa membuat tabel berisi beberapa vektor beserta panjangnya.

```python
vector_list = {
    "a": np.array([1, 2]),
    "b": np.array([3, 4]),
    "c": np.array([5, 12]),
    "d": np.array([-2, 6])
}

rows = []
for name, vec in vector_list.items():
    rows.append({
        "Nama": name,
        "Vektor": str(vec),
        "Panjang": np.linalg.norm(vec)
    })

vector_length_df = pd.DataFrame(rows)
vector_length_df
```

## 20. Vektor dalam Machine Learning

Vektor sangat sering muncul dalam machine learning, misalnya:

1. **Satu sampel data**
   - satu baris data bisa dianggap sebagai vektor fitur

2. **Bobot model**
   - parameter model linear regression atau logistic regression disimpan sebagai vektor bobot

3. **Prediksi**
   - skor prediksi sering diperoleh dari kombinasi linear antara vektor fitur dan vektor bobot

4. **Embedding**
   - kata, gambar, atau item dapat direpresentasikan sebagai vektor numerik

## 21. Contoh Sederhana Vektor Fitur dan Bobot

Misalkan sebuah model menggunakan tiga fitur:
- nilai matematika
- nilai fisika
- nilai pemrograman

Satu siswa direpresentasikan sebagai vektor fitur:
x = [80, 75, 90]

Model memiliki bobot:
w = [0.3, 0.2, 0.5]

Skor prediksi sederhana bisa dihitung sebagai kombinasi elemen-elemen tersebut.
Pada chapter ini kita belum membahas dot product secara formal, tetapi kita bisa melihat idenya melalui perkalian elemen per elemen lalu dijumlahkan.

```python
x = np.array([80, 75, 90])
w = np.array([0.3, 0.2, 0.5])

weighted_values = x * w
prediction_score = np.sum(weighted_values)

result_df = pd.DataFrame({
    "Fitur": ["Matematika", "Fisika", "Pemrograman"],
    "Nilai": x,
    "Bobot": w,
    "Nilai x Bobot": weighted_values
})

print("Skor prediksi sederhana:", round(prediction_score, 2))
result_df
```

## 22. Interpretasi Contoh Prediksi Linear

Dari hasil sebelumnya:
- setiap fitur dikalikan dengan bobotnya
- hasil perkalian dijumlahkan
- diperoleh satu skor akhir

Inilah salah satu ide dasar linear algebra dalam machine learning:
- data direpresentasikan sebagai vektor
- model menyimpan parameter dalam bentuk vektor
- prediksi diperoleh dari operasi antar-vektor
