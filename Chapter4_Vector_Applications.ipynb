# Dasar Teori

## 1. Vektor sebagai Representasi Titik dan Arah
Dalam aljabar linear, vektor dapat dipandang sebagai:
- **arah dan besar**
- **posisi relatif dari titik asal**
- **perpindahan dari satu titik ke titik lain**

Sebagai contoh, vektor:

\[
\mathbf{v} = \begin{bmatrix} 3 \\ 2 \end{bmatrix}
\]

dapat diartikan sebagai:
- bergerak 3 satuan ke arah sumbu-x
- bergerak 2 satuan ke arah sumbu-y

---

## 2. Vektor sebagai Perpindahan (Displacement Vector)
Jika ada dua titik:

\[
A = (x_1, y_1), \quad B = (x_2, y_2)
\]

maka vektor dari \(A\) ke \(B\) adalah:

\[
\vec{AB} = B - A
\]

Vektor ini menunjukkan arah dan besar perpindahan dari titik awal ke titik akhir.

---

## 3. Jarak Antar Titik
Jarak antara dua titik dapat dihitung menggunakan norma dari vektor selisihnya:

\[
\text{distance}(A,B) = ||B-A||
\]

Dalam 2D:

\[
||B-A|| = \sqrt{(x_2-x_1)^2 + (y_2-y_1)^2}
\]

---

## 4. Dot Product untuk Kemiripan Arah
Dot product dua vektor:

\[
\mathbf{u}\cdot\mathbf{v}
\]

dapat digunakan untuk mengetahui seberapa searah dua vektor.  
Jika dot product besar dan positif, maka kedua vektor cenderung searah.  
Jika nol, keduanya tegak lurus.  
Jika negatif, keduanya berlawanan arah.

---

## 5. Proyeksi Vektor
Proyeksi vektor \(\mathbf{u}\) ke arah \(\mathbf{v}\) adalah komponen dari \(\mathbf{u}\) yang berada di sepanjang arah \(\mathbf{v}\):

\[
proj_{\mathbf{v}}(\mathbf{u}) = \frac{\mathbf{u}\cdot\mathbf{v}}{\mathbf{v}\cdot\mathbf{v}} \mathbf{v}
\]

Konsep ini penting dalam:
- dekomposisi vektor
- least squares
- machine learning
- kompresi data
- pemisahan sinyal

---

## 6. Komponen Paralel dan Tegak Lurus
Setiap vektor dapat dipecah menjadi dua bagian:
1. komponen yang **sejajar** dengan vektor acuan
2. komponen yang **tegak lurus** terhadap vektor acuan

Jika \(\mathbf{u}\) diproyeksikan ke \(\mathbf{v}\), maka:

- komponen sejajar = proyeksi \(\mathbf{u}\) pada \(\mathbf{v}\)
- komponen ortogonal = \(\mathbf{u} - proj_{\mathbf{v}}(\mathbf{u})\)

---

## 7. Aplikasi dalam Machine Learning
Konsep vektor sangat penting dalam machine learning karena:
- setiap data numerik bisa dianggap sebagai vektor fitur
- jarak antar data digunakan pada KNN, clustering, dan retrieval
- dot product dipakai dalam model linear, neural network, dan similarity
- proyeksi digunakan dalam PCA, least squares, dan reduksi dimensi

```python
import numpy as np
import matplotlib.pyplot as plt

np.set_printoptions(precision=3, suppress=True)
```

# 1. Vektor sebagai Posisi Titik di Bidang 2D

Kita mulai dengan dua titik di bidang 2D.  
Dari dua titik ini, kita akan melihat bagaimana vektor dapat digunakan untuk merepresentasikan posisi dan perpindahan.

```python
A = np.array([1, 2])
B = np.array([5, 6])

print("Titik A =", A)
print("Titik B =", B)
```

# 2. Vektor Perpindahan dari A ke B

Vektor perpindahan dari titik \(A\) ke titik \(B\) dihitung dengan:

\[
\vec{AB} = B - A
\]

Vektor ini menunjukkan perubahan posisi dari A menuju B.

```python
AB = B - A
print("Vektor AB =", AB)
```

# Analisis Vektor Perpindahan

Jika:

\[
A = (1,2), \quad B = (5,6)
\]

maka:

\[
\vec{AB} = (5-1,\ 6-2) = (4,4)
\]

Artinya, untuk berpindah dari titik A ke titik B, kita perlu bergerak:
- 4 satuan ke kanan
- 4 satuan ke atas

Konsep ini sangat penting dalam geometri, fisika, grafik komputer, dan representasi data.

# 3. Visualisasi Titik A, Titik B, dan Vektor AB

```python
plt.figure(figsize=(7,7))

# Titik A dan B
plt.scatter(A[0], A[1], s=80, label='A')
plt.scatter(B[0], B[1], s=80, label='B')

# Vektor AB dari A ke B
plt.quiver(A[0], A[1], AB[0], AB[1],
           angles='xy', scale_units='xy', scale=1, label='AB')

plt.xlim(0, 7)
plt.ylim(0, 8)
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.grid(True)
plt.legend()
plt.title("Vektor Perpindahan dari A ke B")
plt.xlabel("Sumbu X")
plt.ylabel("Sumbu Y")
plt.show()
```

# Analisis Visualisasi

Pada grafik di atas:
- titik A adalah titik awal
- titik B adalah titik tujuan
- vektor AB menunjukkan arah perpindahan dari A ke B

Hal ini menegaskan bahwa vektor tidak selalu harus digambar dari titik asal `(0,0)`.  
Vektor juga bisa merepresentasikan **perubahan posisi dari satu titik ke titik lain**.

# 4. Menghitung Jarak antara Dua Titik

Jarak antara A dan B dapat dihitung menggunakan norma dari vektor AB:

\[
||AB|| = ||B-A||
\]

```python
distance_AB = np.linalg.norm(AB)
print("Jarak dari A ke B =", distance_AB)
```

# Analisis Jarak

Karena:

\[
AB = (4,4)
\]

maka panjangnya adalah:

\[
||AB|| = \sqrt{4^2 + 4^2} = \sqrt{32}
\]

Jarak ini adalah jarak Euclidean antara titik A dan B.  
Konsep ini sangat penting dalam:
- K-Nearest Neighbors (KNN)
- clustering
- pencarian tetangga terdekat
- pengukuran kemiripan data

# 5. Menggunakan Vektor untuk Mengukur Arah Gerak

Sekarang kita definisikan dua vektor gerak yang berbeda, lalu kita lihat apakah keduanya searah, berlawanan arah, atau membentuk sudut tertentu.

```python
u = np.array([3, 1])
v = np.array([2, 4])

print("u =", u)
print("v =", v)
```

# 6. Dot Product untuk Mengukur Hubungan Arah

```python
dot_uv = np.dot(u, v)
norm_u = np.linalg.norm(u)
norm_v = np.linalg.norm(v)

cos_theta = dot_uv / (norm_u * norm_v)
theta_deg = np.degrees(np.arccos(cos_theta))

print("u · v =", dot_uv)
print("||u|| =", norm_u)
print("||v|| =", norm_v)
print("cos(theta) =", cos_theta)
print("Sudut antara u dan v (derajat) =", theta_deg)
```

# Analisis Hubungan Arah Antar Vektor

Dot product membantu kita memahami apakah dua vektor cenderung searah atau tidak.

Jika hasilnya:
- **positif**, maka sudut antar vektor kurang dari 90°
- **nol**, maka kedua vektor orthogonal
- **negatif**, maka sudut antar vektor lebih dari 90°

Pada kasus ini, nilai dot product positif, sehingga `u` dan `v` memiliki kecenderungan arah yang sama meskipun tidak persis sejajar.

# 7. Visualisasi Dua Vektor dan Sudutnya

```python
plt.figure(figsize=(7,7))

plt.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, label='u')
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, label='v')

plt.xlim(-1, 6)
plt.ylim(-1, 6)
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.grid(True)
plt.legend()
plt.title("Visualisasi Vektor u dan v")
plt.xlabel("Sumbu X")
plt.ylabel("Sumbu Y")
plt.show()
```

# Analisis Visualisasi Arah Vektor

Dari visualisasi terlihat bahwa kedua vektor mengarah ke kuadran yang sama.  
Sudut antara keduanya tidak terlalu besar, sehingga dot product bernilai positif.

Interpretasi seperti ini penting pada:
- cosine similarity
- text mining
- recommendation system
- pembelajaran representasi fitur

# 8. Proyeksi Vektor ke Arah Lain

Sekarang kita proyeksikan vektor `u` ke arah `v`.  
Tujuannya adalah melihat seberapa besar komponen `u` yang “jatuh” ke arah `v`.

```python
proj_u_on_v = (np.dot(u, v) / np.dot(v, v)) * v
print("Proyeksi u pada v =", proj_u_on_v)
```

# Analisis Proyeksi

Proyeksi `u` pada `v` adalah bagian dari `u` yang searah dengan `v`.  
Jika kita bayangkan `u` sebagai panah dan `v` sebagai arah referensi, maka proyeksi adalah “bayangan” `u` di atas arah `v`.

Konsep ini digunakan pada:
- least squares
- dekomposisi sinyal
- analisis komponen
- representasi fitur

# 9. Menghitung Komponen Tegak Lurus terhadap Vektor Acuan

Setelah menghitung komponen sejajar (proyeksi), kita bisa menghitung komponen yang tegak lurus terhadap `v`:

\[
u_{\perp} = u - proj_v(u)
\]

```python
u_perp = u - proj_u_on_v
print("Komponen orthogonal u terhadap v =", u_perp)
```

# Analisis Komponen Tegak Lurus

Sekarang vektor `u` telah dipecah menjadi dua bagian:
1. **komponen sejajar dengan `v`** → `proj_u_on_v`
2. **komponen orthogonal terhadap `v`** → `u_perp`

Secara matematis:

\[
u = proj_v(u) + u_{\perp}
\]

Dekomposisi seperti ini sangat penting dalam banyak metode numerik dan machine learning.

# 10. Verifikasi Bahwa Komponen Orthogonal Benar-Benar Tegak Lurus

Jika `u_perp` benar-benar orthogonal terhadap `v`, maka dot product-nya dengan `v` harus mendekati nol.

```python
check_orthogonal = np.dot(u_perp, v)
print("u_perp · v =", check_orthogonal)
```

# Analisis Verifikasi Orthogonalitas

Jika hasil dot product sangat kecil atau mendekati nol, maka komponen tersebut benar-benar orthogonal terhadap `v`.

Ini adalah salah satu ide paling penting dalam aljabar linear:
- suatu vektor dapat dipisah menjadi komponen searah suatu subruang
- dan komponen sisanya yang orthogonal terhadap subruang tersebut

# 11. Visualisasi Dekomposisi Vektor

Kita akan menampilkan:
- vektor asli `u`
- vektor acuan `v`
- proyeksi `u` pada `v`
- komponen orthogonal `u_perp`

```python
plt.figure(figsize=(8,8))

# vektor asli u
plt.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, label='u')

# vektor acuan v
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, label='v')

# proyeksi u pada v
plt.quiver(0, 0, proj_u_on_v[0], proj_u_on_v[1],
           angles='xy', scale_units='xy', scale=1, label='proj_u_on_v')

# komponen orthogonal dimulai dari ujung proyeksi
plt.quiver(proj_u_on_v[0], proj_u_on_v[1], u_perp[0], u_perp[1],
           angles='xy', scale_units='xy', scale=1, label='u_perp')

plt.xlim(-1, 6)
plt.ylim(-1, 6)
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.grid(True)
plt.legend()
plt.title("Dekomposisi Vektor menjadi Komponen Sejajar dan Orthogonal")
plt.xlabel("Sumbu X")
plt.ylabel("Sumbu Y")
plt.show()
```

# Analisis Visualisasi Dekomposisi

Dari gambar terlihat bahwa:
- `proj_u_on_v` berada di sepanjang arah `v`
- `u_perp` merupakan bagian sisa dari `u` yang tegak lurus terhadap `v`
- jika dijumlahkan, keduanya membentuk kembali vektor `u`

Visualisasi ini sangat penting karena memberikan intuisi geometris terhadap konsep proyeksi dan orthogonalitas.

# 12. Aplikasi Vektor pada Kemiripan Data

Dalam machine learning, setiap observasi sering direpresentasikan sebagai vektor fitur.  
Kita bisa mengukur kemiripan dua data dengan **cosine similarity**, yaitu:

\[
\text{cosine similarity} = \frac{u \cdot v}{||u|| \ ||v||}
\]

Semakin mendekati 1, semakin mirip arah kedua vektor.

```python
x1 = np.array([1, 2, 3, 4])
x2 = np.array([2, 4, 6, 8])
x3 = np.array([4, 1, 0, 2])

cos_12 = np.dot(x1, x2) / (np.linalg.norm(x1) * np.linalg.norm(x2))
cos_13 = np.dot(x1, x3) / (np.linalg.norm(x1) * np.linalg.norm(x3))

print("Cosine similarity x1 dan x2 =", cos_12)
print("Cosine similarity x1 dan x3 =", cos_13)
```

# Analisis Cosine Similarity

Vektor `x1` dan `x2` memiliki pola yang sangat mirip karena `x2` adalah kelipatan dari `x1`.  
Akibatnya, cosine similarity antara keduanya mendekati 1.

Sebaliknya, `x1` dan `x3` memiliki pola yang berbeda sehingga cosine similarity-nya lebih kecil.

Konsep ini sangat sering digunakan dalam:
- text classification
- information retrieval
- recommendation system
- embedding similarity

# 13. Aplikasi Vektor pada Jarak Data

Selain kemiripan arah, kita juga sering perlu mengukur jarak antar data.  
Sebagai contoh, kita hitung jarak Euclidean antara dua observasi fitur.

```python
dist_13 = np.linalg.norm(x1 - x3)
dist_12 = np.linalg.norm(x1 - x2)

print("Jarak Euclidean x1 ke x2 =", dist_12)
print("Jarak Euclidean x1 ke x3 =", dist_13)
```

# Analisis Jarak Data

Jarak Euclidean mengukur seberapa jauh dua titik dalam ruang fitur.  
Berbeda dengan cosine similarity yang fokus pada arah, jarak Euclidean fokus pada **selisih absolut posisi**.

Keduanya sama-sama penting, tetapi dipakai pada konteks yang berbeda:
- **Euclidean distance** → KNN, clustering, nearest neighbors
- **cosine similarity** → dokumen teks, embedding, similarity search

# 14. Studi Mini: Menentukan Titik yang Paling Mirip terhadap Query

Misalkan kita punya satu vektor query dan beberapa kandidat data.  
Kita ingin melihat data mana yang paling mirip dengan query berdasarkan cosine similarity.

```python
query = np.array([1, 2, 1])

data = np.array([
    [1, 2, 1],
    [2, 4, 2],
    [0, 1, 3],
    [3, 1, 0]
])

def cosine_sim(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))

scores = []
for i, row in enumerate(data):
    score = cosine_sim(query, row)
    scores.append(score)
    print(f"Data ke-{i+1}: {row}, cosine similarity = {score:.4f}")

best_idx = np.argmax(scores)
print("\nData paling mirip dengan query adalah data ke-", best_idx+1)
print("Vektornya:", data[best_idx])
```

# Analisis Studi Mini

Pada contoh ini, query dibandingkan dengan setiap data menggunakan cosine similarity.  
Nilai similarity tertinggi menunjukkan bahwa arah vektor kandidat paling mirip dengan query.

Pendekatan seperti ini adalah dasar dari:
- pencarian dokumen mirip
- recommendation system
- semantic search
- retrieval berbasis embedding
