## 2. Apa Itu Matriks?

Matriks adalah susunan angka berbentuk persegi panjang yang terdiri dari:
- **baris (rows)**
- **kolom (columns)**

Contoh matriks 2×3:

[[1, 2, 3],
 [4, 5, 6]]

Artinya:
- memiliki 2 baris
- memiliki 3 kolom

Dalam data science:
- **setiap baris** sering merepresentasikan satu observasi / satu sampel data
- **setiap kolom** merepresentasikan satu fitur / variabel

## 3. Membuat Matriks dengan NumPy

Kita dapat membuat matriks menggunakan `np.array()` dengan memasukkan list di dalam list.

```python
A = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

B = np.array([
    [10, 20],
    [30, 40],
    [50, 60]
])

print("Matriks A:")
print(A)

print("\nMatriks B:")
print(B)
```

## 4. Shape, Dimensi, dan Ukuran Matriks

Beberapa atribut penting matriks:
- `shape` → ukuran matriks (baris, kolom)
- `ndim` → jumlah dimensi array
- `size` → total jumlah elemen

Untuk matriks, `shape` biasanya ditulis `(jumlah_baris, jumlah_kolom)`.

```python
matrix_info = pd.DataFrame({
    "Nama Matriks": ["A", "B"],
    "Shape": [A.shape, B.shape],
    "Jumlah Dimensi": [A.ndim, B.ndim],
    "Jumlah Elemen": [A.size, B.size]
})

matrix_info
```

## 5. Menampilkan Matriks sebagai Tabel

Agar lebih mudah dibaca, matriks bisa kita tampilkan sebagai DataFrame pandas.

```python
A_df = pd.DataFrame(A, columns=["Kolom_1", "Kolom_2", "Kolom_3"])
B_df = pd.DataFrame(B, columns=["Kolom_1", "Kolom_2"])

print("DataFrame untuk matriks A:")
display(A_df)

print("DataFrame untuk matriks B:")
display(B_df)
```

## 6. Mengakses Elemen Matriks

Karena matriks memiliki baris dan kolom, elemen matriks diakses dengan dua indeks:
`matriks[baris, kolom]`

Contoh:
- `A[0,0]` → elemen baris 1 kolom 1
- `A[1,2]` → elemen baris 2 kolom 3

Ingat bahwa indeks Python dimulai dari 0.

```python
print("Matriks A:")
print(A)

print("\nElemen A[0,0] =", A[0,0])
print("Elemen A[0,1] =", A[0,1])
print("Elemen A[1,2] =", A[1,2])
```

## 7. Mengambil Satu Baris atau Satu Kolom

Pada matriks:
- satu **baris** bisa diambil dengan `A[i, :]`
- satu **kolom** bisa diambil dengan `A[:, j]`

Ini penting dalam data science karena:
- baris = satu observasi
- kolom = satu fitur

```python
row_1 = A[0, :]
row_2 = A[1, :]
col_1 = A[:, 0]
col_2 = A[:, 1]

print("Baris pertama A:", row_1)
print("Baris kedua A :", row_2)
print("Kolom pertama A:", col_1)
print("Kolom kedua A :", col_2)
```

## 8. Memahami Matriks sebagai Kumpulan Vektor

Matriks dapat dipandang sebagai kumpulan vektor:
- setiap **baris** adalah row vector
- setiap **kolom** adalah column vector

Contoh pada dataset:
- 100 baris = 100 sampel
- 4 kolom = 4 fitur

```python
sample_data = np.array([
    [170, 65, 22],
    [160, 55, 20],
    [180, 75, 24],
    [175, 70, 23]
])

sample_df = pd.DataFrame(
    sample_data,
    columns=["Tinggi", "Berat", "Usia"],
    index=[f"Sampel_{i}" for i in range(1, 5)]
)

sample_df
```

## 9. Matriks Nol, Matriks Satu, dan Matriks Identitas

Beberapa jenis matriks dasar yang sering digunakan:

1. **Matriks nol**
   - semua elemennya 0

2. **Matriks satu**
   - semua elemennya 1

3. **Matriks identitas**
   - diagonal utama bernilai 1
   - elemen lain bernilai 0

Matriks identitas sangat penting dalam aljabar linear.

```python
zero_matrix = np.zeros((3, 4))
ones_matrix = np.ones((3, 4))
identity_matrix = np.eye(4)

print("Matriks nol:")
print(zero_matrix)

print("\nMatriks satu:")
print(ones_matrix)

print("\nMatriks identitas:")
print(identity_matrix)
```

## 10. Membuat Matriks Acak

Dalam eksperimen data science, kita sering membutuhkan matriks acak untuk simulasi.

## 11. Penjumlahan Matriks

Dua matriks dapat dijumlahkan jika ukurannya sama.

Misalnya:
A dan B sama-sama berukuran 2×3

Maka:
A + B dilakukan elemen per elemen.

```python
M1 = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

M2 = np.array([
    [10, 20, 30],
    [40, 50, 60]
])

M_add = M1 + M2

print("M1:")
print(M1)

print("\nM2:")
print(M2)

print("\nM1 + M2:")
print(M_add)
```

## 12. Pengurangan Matriks

Pengurangan matriks juga dilakukan elemen per elemen, asalkan ukuran kedua matriks sama.

```python
M_sub = M2 - M1

print("M2 - M1:")
print(M_sub)
```

## 13. Perkalian Skalar pada Matriks

Perkalian skalar berarti mengalikan semua elemen matriks dengan satu angka.

```python
scalar = 3
M_scaled = scalar * M1

print("M1:")
print(M1)

print(f"\n{scalar} * M1:")
print(M_scaled)
```

## 14. Operasi Elemen per Elemen pada Matriks

Selain penjumlahan dan pengurangan, NumPy juga memungkinkan operasi elemen per elemen seperti:
- perkalian elemen per elemen
- pembagian elemen per elemen
- pemangkatan elemen per elemen

```python
elem_mult = M1 * M2
elem_div = M2 / M1
elem_power = M1 ** 2

print("Perkalian elemen per elemen:")
print(elem_mult)

print("\nPembagian elemen per elemen:")
print(elem_div)

print("\nPangkat dua setiap elemen M1:")
print(elem_power)
```

## 15. Transpose Matriks

Transpose adalah operasi menukar baris menjadi kolom, dan kolom menjadi baris.

Jika matriks A berukuran `(m, n)`, maka transpose A akan berukuran `(n, m)`.

Notasi transpose biasanya ditulis:
Aᵀ

```python
A = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

A_T = A.T

print("Matriks A:")
print(A)
print("Shape A:", A.shape)

print("\nTranspose A:")
print(A_T)
print("Shape A.T:", A_T.shape)
```

## 16. Memahami Transpose Secara Visual

Transpose akan:
- mengubah kolom pertama menjadi baris pertama
- kolom kedua menjadi baris kedua
- dan seterusnya

```python
A_df = pd.DataFrame(A, columns=["C1", "C2", "C3"])
AT_df = pd.DataFrame(A_T, columns=["R1", "R2"])

print("Matriks A:")
display(A_df)

print("Transpose A:")
display(AT_df)
```

## 17. Sifat Dasar Transpose

Beberapa sifat dasar transpose:
1. `(A^T)^T = A`
2. `transpose` tidak mengubah jumlah elemen
3. ukuran matriks berubah dari `(m,n)` menjadi `(n,m)`

```python
double_transpose = A.T.T

print("A:")
print(A)

print("\n(A.T).T:")
print(double_transpose)

print("\nApakah sama dengan A?", np.array_equal(A, double_transpose))
```

## 18. Matriks Simetris

Matriks disebut **simetris** jika:
A = Aᵀ

Syarat penting:
- matriks harus persegi (jumlah baris = jumlah kolom)

```python
S = np.array([
    [1, 2, 3],
    [2, 5, 6],
    [3, 6, 9]
])

print("Matriks S:")
print(S)

print("\nTranspose S:")
print(S.T)

print("\nApakah S simetris?", np.array_equal(S, S.T))
```

## 19. Matriks Diagonal

Matriks diagonal adalah matriks persegi yang hanya memiliki nilai bukan nol pada diagonal utamanya.

```python
D = np.diag([5, 10, 15, 20])

print("Matriks diagonal:")
print(D)
```

## 20. Matriks dalam Data Science

Dalam machine learning, dataset tabular hampir selalu dapat dipandang sebagai matriks.

Contoh:
- 150 baris data Iris
- 4 kolom fitur numerik

Maka dataset fitur dapat dianggap sebagai matriks 150×4.

```python
students = np.array([
    [80, 75, 90],
    [70, 85, 88],
    [92, 78, 95],
    [65, 70, 72],
    [88, 90, 84]
])

students_df = pd.DataFrame(
    students,
    columns=["Matematika", "Fisika", "Pemrograman"],
    index=[f"Siswa_{i}" for i in range(1, 6)]
)

students_df
```

## 21. Setiap Baris sebagai Satu Sampel, Setiap Kolom sebagai Satu Fitur

Dari matriks data siswa:
- satu baris = satu siswa
- satu kolom = satu mata pelajaran

Kita bisa mengambil:
- data satu siswa
- seluruh nilai satu mata pelajaran

```python
siswa_1 = students[0, :]
nilai_matematika = students[:, 0]

print("Data siswa pertama:", siswa_1)
print("Semua nilai matematika:", nilai_matematika)
```

## 22. Rata-rata Tiap Kolom

Dalam data science, kita sering menghitung statistik per fitur.  
Contohnya rata-rata tiap kolom.

```python
column_means = students.mean(axis=0)

mean_df = pd.DataFrame({
    "Fitur": ["Matematika", "Fisika", "Pemrograman"],
    "Rata-rata": column_means
})

mean_df
```

## 23. Rata-rata Tiap Baris

Sebaliknya, kita juga bisa menghitung rata-rata per baris, misalnya rata-rata nilai tiap siswa.

```python
row_means = students.mean(axis=1)

row_mean_df = pd.DataFrame({
    "Siswa": [f"Siswa_{i}" for i in range(1, 6)],
    "Rata-rata Nilai": row_means
})

row_mean_df
```

## 24. Menjumlahkan Matriks per Baris dan per Kolom

Operasi agregasi ini sering dipakai saat eksplorasi data:
- `axis=0` → operasi per kolom
- `axis=1` → operasi per baris

```python
sum_per_column = students.sum(axis=0)
sum_per_row = students.sum(axis=1)

print("Jumlah per kolom:", sum_per_column)
print("Jumlah per baris:", sum_per_row)
```

## 25. Visualisasi Matriks sebagai Heatmap

Salah satu cara sederhana untuk melihat pola nilai dalam matriks adalah menggunakan heatmap.

```python
plt.figure(figsize=(8, 5))
sns.heatmap(students_df, annot=True, cmap="YlGnBu", fmt="d")
plt.title("Heatmap Matriks Nilai Siswa")
plt.show()
```

## 26. Heatmap untuk Matriks Acak

Kita juga bisa memvisualisasikan matriks acak untuk melihat perbedaan intensitas nilai.

```python
np.random.seed(7)
heatmap_matrix = np.random.randint(1, 100, size=(6, 6))

plt.figure(figsize=(7, 5))
sns.heatmap(heatmap_matrix, annot=True, cmap="viridis", fmt="d")
plt.title("Heatmap Matriks Acak")
plt.show()
```

## 28. Contoh Matriks sebagai Dataset Fitur

- seluruh luas rumah
