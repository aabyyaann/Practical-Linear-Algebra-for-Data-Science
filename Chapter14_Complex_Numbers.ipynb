# Ringkasan Teori

Bilangan kompleks ditulis sebagai:

\[
z = a + bi
\]

dimana:

- a = bagian real
- b = bagian imajiner

dan:

\[
i=\sqrt{-1}
\]

Karena:

\[
i^2=-1
\]

Bilangan kompleks memungkinkan kita menyelesaikan masalah yang tidak dapat diselesaikan menggunakan bilangan real saja.

````python
import numpy as np
import matplotlib.pyplot as plt

np.set_printoptions(precision=4, suppress=True)
````

# 1. Membuat Bilangan Kompleks

Python menggunakan huruf:

```python
j
```

sebagai pengganti:

\[
i
\]

dalam matematika.

````python
z = 3 + 4j

print(z)
````

# 2. Bagian Real dan Imajiner

Bilangan kompleks terdiri dari:

\[
z=a+bi
\]

Kita dapat mengambil masing-masing komponennya.

````python
print("Real part:")
print(z.real)

print("\nImaginary part:")
print(z.imag)
````

# 3. Penjumlahan Bilangan Kompleks

Jika:

\[
z_1=a+bi
\]

\[
z_2=c+di
\]

maka:

\[
z_1+z_2=(a+c)+(b+d)i
\]

````python
z1 = 2 + 3j
z2 = 4 + 5j

print(z1 + z2)
````

# 4. Pengurangan Bilangan Kompleks

````python
print(z1 - z2)
````

# 5. Perkalian Bilangan Kompleks

Perkalian mengikuti aturan:

\[
i^2=-1
\]

````python
print(z1 * z2)
````

# 6. Pembagian Bilangan Kompleks

````python
print(z1 / z2)
````

# 7. Konjugat Kompleks

Konjugat kompleks diperoleh dengan mengganti tanda bagian imajiner.

Jika:

\[
z=a+bi
\]

maka:

\[
z^*=a-bi
\]

````python
z = 3 + 4j

print(z.conjugate())
````

# Mengapa Konjugat Penting?

Konjugat digunakan dalam:

- Hermitian transpose
- Fourier Transform
- Signal Processing
- Quantum Computing

````python
z = 3 + 4j

print(z * z.conjugate())
````

# Hasil

Secara teori:

\[
(a+bi)(a-bi)
=
a^2+b^2
\]

yang selalu menghasilkan bilangan real.

# 8. Magnitude Bilangan Kompleks

Magnitude atau panjang bilangan kompleks:

\[
|z|
=
\sqrt{a^2+b^2}
\]

````python
z = 3 + 4j

print(abs(z))
````

# Verifikasi Manual

````python
manual_mag = np.sqrt(
    z.real**2 +
    z.imag**2
)

print(manual_mag)
````

# 9. Visualisasi Bilangan Kompleks

Bilangan kompleks dapat direpresentasikan sebagai titik pada bidang kompleks.

Sumbu:

- x = real
- y = imajiner

````python
z = 3 + 4j

plt.figure(figsize=(6,6))

plt.scatter(
    z.real,
    z.imag
)

plt.axhline(0)
plt.axvline(0)

plt.xlabel("Real")
plt.ylabel("Imaginary")

plt.grid(True)
plt.axis('equal')

plt.show()
````

# 10. Sudut (Phase)

Selain magnitude, bilangan kompleks memiliki sudut:

\[
\theta
\]

yang disebut phase.

````python
theta = np.angle(z)

print(theta)
````

# Phase dalam Derajat

````python
print(np.degrees(theta))
````

# 11. Bentuk Polar

Bilangan kompleks dapat ditulis sebagai:

\[
z=r(\cos\theta+i\sin\theta)
\]

dimana:

- r = magnitude
- θ = phase

````python
r = abs(z)

theta = np.angle(z)

print("Magnitude =", r)
print("Angle =", theta)
````

# Rekonstruksi dari Bentuk Polar

````python
z_reconstructed = r * (
    np.cos(theta)
    +
    1j*np.sin(theta)
)

print(z_reconstructed)
````

# 12. Euler's Formula

Salah satu persamaan paling terkenal dalam matematika:

\[
e^{i\theta}
=
\cos\theta+i\sin\theta
\]

Persamaan ini menghubungkan:

- eksponensial
- trigonometri
- bilangan kompleks

````python
theta = np.pi/4

left = np.exp(1j*theta)

right = (
    np.cos(theta)
    +
    1j*np.sin(theta)
)

print(left)
print(right)
````

# Verifikasi Euler Formula

````python
np.allclose(left,right)
````

# 13. Euler Identity

Kasus khusus:

\[
e^{i\pi}
=
-1
\]

sehingga:

\[
e^{i\pi}+1=0
\]

Persamaan ini dikenal sebagai Euler Identity.

````python
np.exp(1j*np.pi)
````

````python
np.exp(1j*np.pi) + 1
````

# 14. Visualisasi Rotasi Kompleks

Perkalian dengan:

\[
e^{i\theta}
\]

menghasilkan rotasi pada bidang kompleks.

````python
z = 1 + 0j

theta = np.pi/3

rotated = z * np.exp(1j*theta)

print(rotated)
````

````python
plt.figure(figsize=(6,6))

plt.quiver(
    0,0,
    z.real,z.imag,
    angles='xy',
    scale_units='xy',
    scale=1
)

plt.quiver(
    0,0,
    rotated.real,
    rotated.imag,
    angles='xy',
    scale_units='xy',
    scale=1
)

plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)

plt.grid(True)
plt.axis('equal')

plt.show()
````
