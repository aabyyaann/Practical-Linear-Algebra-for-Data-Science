# Dasar Teori

## 1. Penjumlahan Vektor
Jika terdapat dua vektor dengan dimensi yang sama:

\[
\mathbf{u} = [u_1, u_2, \dots, u_n]
\]

\[
\mathbf{v} = [v_1, v_2, \dots, v_n]
\]

maka hasil penjumlahannya adalah:

\[
\mathbf{u} + \mathbf{v} = [u_1+v_1, u_2+v_2, \dots, u_n+v_n]
\]

Penjumlahan vektor dilakukan elemen per elemen.

## 2. Pengurangan Vektor
Pengurangan vektor juga dilakukan per elemen:

\[
\mathbf{u} - \mathbf{v} = [u_1-v_1, u_2-v_2, \dots, u_n-v_n]
\]

## 3. Perkalian Skalar
Jika sebuah vektor dikalikan dengan bilangan skalar \(c\), maka setiap elemen vektor dikalikan dengan \(c\):

\[
c\mathbf{u} = [cu_1, cu_2, \dots, cu_n]
\]

Operasi ini akan mengubah panjang vektor, dan jika nilai skalar negatif maka arah vektor juga berbalik.

## 4. Norma Vektor
Norma vektor menyatakan panjang atau besar suatu vektor. Norma yang paling umum adalah **Euclidean norm**:

\[
||\mathbf{u}|| = \sqrt{u_1^2 + u_2^2 + \dots + u_n^2}
\]

Norma digunakan untuk mengukur seberapa besar sebuah vektor.


## 5. Dot Product
Dot product atau perkalian titik dari dua vektor adalah:

\[
\mathbf{u} \cdot \mathbf{v} = u_1v_1 + u_2v_2 + \dots + u_nv_n
\]

Dot product sangat penting karena digunakan untuk:
- mengukur kemiripan arah dua vektor
- menghitung sudut antar vektor
- proyeksi vektor
- perhitungan pada machine learning dan neural network


## 6. Sudut Antar Vektor
Sudut antara dua vektor dapat dihitung dengan rumus:

\[
\cos(\theta) = \frac{\mathbf{u}\cdot\mathbf{v}}{||\mathbf{u}|| \ ||\mathbf{v}||}
\]

Sehingga:

\[
\theta = \cos^{-1}\left(\frac{\mathbf{u}\cdot\mathbf{v}}{||\mathbf{u}|| \ ||\mathbf{v}||}\right)
\]

Jika dot product bernilai:
- positif → sudut cenderung lancip
- nol → vektor saling tegak lurus
- negatif → sudut tumpul


## 7. Proyeksi Vektor
Proyeksi vektor \(\mathbf{u}\) ke vektor \(\mathbf{v}\) adalah:

\[
proj_{\mathbf{v}}(\mathbf{u}) = \frac{\mathbf{u}\cdot\mathbf{v}}{\mathbf{v}\cdot\mathbf{v}} \mathbf{v}
\]

Konsep ini digunakan untuk memecah vektor ke arah tertentu, dan sangat penting dalam optimasi, regresi, serta dekomposisi data.

```python
# Import library utama
import numpy as np
import matplotlib.pyplot as plt

# Supaya angka desimal lebih rapi
np.set_printoptions(precision=3, suppress=True)
```

# 1. Membuat Vektor

Pada bagian awal, kita definisikan dua vektor sederhana untuk digunakan pada seluruh eksperimen di chapter ini.

```python
u = np.array([2, 3])
v = np.array([1, 4])

print("Vektor u:", u)
print("Vektor v:", v)
```

# 2. Penjumlahan dan Pengurangan Vektor

Operasi pertama yang akan dicoba adalah penjumlahan dan pengurangan vektor. Karena kedua vektor memiliki dimensi yang sama, maka operasi ini dapat dilakukan secara langsung.

```python
u_plus_v = u + v
u_minus_v = u - v

print("u + v =", u_plus_v)
print("u - v =", u_minus_v)
```

# Analisis Hasil Penjumlahan dan Pengurangan

Hasil penjumlahan menunjukkan bahwa setiap elemen pada vektor `u` dijumlahkan dengan elemen yang bersesuaian pada vektor `v`.

Sebagai contoh:
- elemen pertama: \(2 + 1 = 3\)
- elemen kedua: \(3 + 4 = 7\)

Pengurangan juga bekerja dengan prinsip yang sama:
- elemen pertama: \(2 - 1 = 1\)
- elemen kedua: \(3 - 4 = -1\)

Hal ini menunjukkan bahwa operasi vektor di NumPy sangat dekat dengan definisi matematisnya.

# 3. Perkalian Skalar pada Vektor

Sekarang kita kalikan vektor dengan suatu bilangan skalar. Operasi ini berguna untuk mengubah panjang vektor tanpa mengubah arah (jika skalarnya positif).

```python
scalar = 3
scaled_u = scalar * u

print("Skalar:", scalar)
print("Hasil 3 * u =", scaled_u)
```

```python
negative_scalar = -2
scaled_v = negative_scalar * v

print("Skalar:", negative_scalar)
print("Hasil -2 * v =", scaled_v)
```

# Analisis Perkalian Skalar

Ketika vektor `u = [2, 3]` dikalikan dengan 3, maka hasilnya menjadi:

\[
3[2,3] = [6,9]
\]

Sedangkan ketika vektor `v = [1,4]` dikalikan dengan -2, hasilnya:

\[
-2[1,4] = [-2,-8]
\]

Skalar negatif tidak hanya memperbesar panjang vektor, tetapi juga membalik arah vektor terhadap titik asal.

# 4. Visualisasi Vektor Dasar

Agar lebih mudah dipahami, kita visualisasikan vektor `u` dan `v` dalam bidang 2 dimensi.

```python
plt.figure(figsize=(6, 6))

# gambar vektor u
plt.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, label='u')

# gambar vektor v
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, label='v')

plt.xlim(-3, 6)
plt.ylim(-3, 6)
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.grid(True)
plt.legend()
plt.title("Visualisasi Vektor u dan v")
plt.xlabel("Sumbu X")
plt.ylabel("Sumbu Y")
plt.show()
```

# Analisis Visualisasi

Dari visualisasi tersebut:
- vektor `u = [2,3]` bergerak 2 langkah ke kanan dan 3 langkah ke atas
- vektor `v = [1,4]` bergerak 1 langkah ke kanan dan 4 langkah ke atas

Setiap vektor digambar dari titik asal `(0,0)`. Visualisasi ini membantu memahami arah dan besar vektor secara geometris.

# 5. Menghitung Panjang Vektor (Norma)

Norma vektor menunjukkan panjang vektor. Pada NumPy, kita bisa menghitungnya menggunakan `np.linalg.norm()`.

```python
norm_u = np.linalg.norm(u)
norm_v = np.linalg.norm(v)

print("Norma u =", norm_u)
print("Norma v =", norm_v)
```

# Analisis Norma Vektor

Norma vektor `u = [2,3]` dihitung sebagai:

\[
||u|| = \sqrt{2^2 + 3^2} = \sqrt{13}
\]

Sedangkan norma vektor `v = [1,4]` adalah:

\[
||v|| = \sqrt{1^2 + 4^2} = \sqrt{17}
\]

Nilai ini merepresentasikan panjang vektor dari titik asal ke ujung vektor.

# 6. Dot Product

Dot product adalah salah satu operasi paling penting dalam aljabar linear. Operasi ini mengalikan elemen yang bersesuaian, lalu menjumlahkannya.

```python
dot_uv = np.dot(u, v)

print("u =", u)
print("v =", v)
print("u · v =", dot_uv)
```

# Analisis Dot Product

Dengan:

\[
u = [2,3], \quad v = [1,4]
\]

maka:

\[
u \cdot v = (2)(1) + (3)(4) = 2 + 12 = 14
\]

Nilai dot product cukup besar dan positif, yang menunjukkan bahwa kedua vektor memiliki arah yang cukup sejalan.

# 7. Sudut Antar Vektor

Dot product juga bisa digunakan untuk menghitung sudut antara dua vektor.

```python
cos_theta = np.dot(u, v) / (np.linalg.norm(u) * np.linalg.norm(v))
theta_rad = np.arccos(cos_theta)
theta_deg = np.degrees(theta_rad)

print("cos(theta) =", cos_theta)
print("Sudut dalam radian =", theta_rad)
print("Sudut dalam derajat =", theta_deg)
```

# Analisis Sudut Antar Vektor

Jika nilai cosinus mendekati 1, maka sudut antar vektor kecil dan kedua vektor cenderung searah.  
Pada hasil di atas, sudut antara `u` dan `v` bernilai kurang dari 90 derajat, sehingga keduanya membentuk sudut lancip.

Konsep sudut antar vektor sering digunakan dalam:
- cosine similarity pada text mining
- pengukuran kemiripan antar data
- sistem rekomendasi
- pencarian dokumen

# 8. Proyeksi Vektor

Sekarang kita hitung proyeksi vektor `u` ke arah `v`.

```python
proj_u_on_v = (np.dot(u, v) / np.dot(v, v)) * v

print("Proyeksi u pada v =", proj_u_on_v)
```

# Analisis Proyeksi Vektor

Proyeksi menunjukkan bagian dari vektor `u` yang searah dengan `v`.  
Secara intuitif, proyeksi menjawab pertanyaan:

**“Jika vektor `u` dijatuhkan ke arah `v`, berapa besar komponen `u` yang berada di sepanjang `v`?”**

Konsep ini sangat penting dalam:
- regresi linear
- dekomposisi data
- kompresi informasi
- optimasi parameter model

# 9. Visualisasi Proyeksi Vektor

Agar lebih mudah dipahami, kita visualisasikan vektor `u`, `v`, dan hasil proyeksi `u` pada `v`.

```python
plt.figure(figsize=(7, 7))

# vektor u
plt.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, label='u')

# vektor v
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, label='v')

# proyeksi u pada v
plt.quiver(0, 0, proj_u_on_v[0], proj_u_on_v[1],
           angles='xy', scale_units='xy', scale=1, label='proj_u_on_v')

plt.xlim(-1, 5)
plt.ylim(-1, 6)
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.grid(True)
plt.legend()
plt.title("Visualisasi Proyeksi Vektor")
plt.xlabel("Sumbu X")
plt.ylabel("Sumbu Y")
plt.show()
```

# Analisis Visualisasi Proyeksi

Dari gambar, vektor proyeksi berada pada arah yang sama dengan `v`. Hal ini menunjukkan bahwa proyeksi memang merupakan “bayangan” dari `u` di sepanjang arah `v`.

Semakin besar dot product antara `u` dan `v`, semakin besar pula komponen `u` yang jatuh ke arah `v`.

# 10. Membandingkan Beberapa Vektor

Sekarang kita coba beberapa pasangan vektor untuk melihat hubungan antara dot product dan sudut.

```python
a = np.array([1, 0])
b = np.array([0, 1])
c = np.array([2, 2])

print("a =", a)
print("b =", b)
print("c =", c)

print("\na · b =", np.dot(a, b))
print("a · c =", np.dot(a, c))
print("b · c =", np.dot(b, c))
```

# Analisis Hubungan Dot Product

- `a · b = 0` menunjukkan bahwa `a` dan `b` saling tegak lurus (orthogonal)
- `a · c > 0` menunjukkan bahwa `a` dan `c` memiliki komponen arah yang sejalan
- `b · c > 0` juga menunjukkan hal serupa

Dari sini terlihat bahwa dot product bukan hanya operasi aritmetika, tetapi juga alat untuk memahami hubungan geometris antar vektor.

# 11. Eksperimen Tambahan: Vektor Orthogonal

Dua vektor dikatakan orthogonal jika dot product-nya bernilai nol.

```python
p = np.array([3, -3])
q = np.array([2, 2])

dot_pq = np.dot(p, q)

print("p =", p)
print("q =", q)
print("p · q =", dot_pq)
```

# Analisis Orthogonalitas

Hasil dot product `p · q = 0` menunjukkan bahwa kedua vektor orthogonal.  
Dalam geometri, artinya sudut antara keduanya adalah 90 derajat.

Konsep orthogonalitas sangat penting dalam:
- basis ortogonal
- proyeksi
- PCA
- dekomposisi matriks
- machine learning berbasis ruang vektor
